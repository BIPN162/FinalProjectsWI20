## Introduction

**Abstract**

In the project, we investigate RORb+ neurons in visual cortex through electrophysiology, visual stimuli response, and neuronal connectivity in order to characterize their physical properties and to determine whether we can identify subtypes of RORb+ neurons in open datasets. Electrophysiological analysis revealed that RORb+ neurons are less excitable, and failed to reveal novel subtypes. Analyzing response to visual stimulus identified changes in behavior across visual subregions but still failed to identify a consistent subtype. Connectivity data has identified a potential subtype of RORb+ neurons in lateral visual cortex that project to other cortical areas related to the spatial recognition, which shed a light to the future study on the RORb+ function in development of such pathways. 


**Research Question**

In this project, we sought to characterize properties of RORb+ neurons and to identify subpopulations of RORb+ neurons based on electrophysiological properties, response to visual stimuli, and neuronal connectivity. 

**Background**

Cortical layer 4 is comprised of spiny stellate cells that receive sensory input from the thalamus [1]. Layer 4 cortical neurons preferentially express certain marker genes, one of which is RAR orphan receptor beta (RORb) [1]. As a type of excitatory neurons in cortical area, RORb+ cells are involved in the recurrent excitatory circuitry and contribute to the development of the neocortex cortical layers [5, 10]. In this project, we seek to characterize the properties and subtypes of RORb+ neurons in the visual cortex based on their electrophysiology, visual response and its connectivity. 

Electrophysiological properties were obtained from the Allen Cell Types Database, in which mouse transgenically-labelled, reporter-positive neurons are recorded by patch clamp electrophysiology in vitro, in response to injected current stimuli [2]. The dataset includes metrics related to the time and voltage of single AP waveform landmarks, spike train properties (eg. latency, average interspike interval, FI curve slope), and membrane electrical constants [2]. There are 1585 recordings for all neurons in visual areas, and among them 151 are from RORb+ neurons.

Visual response properties were obtained from the Allen Brain Observatory Database. Mice transgenically expressing a calcium sensor (GCaMP6) under a gene promoter were subjected to various visual stimuli (including gratings and images) Neuronal response was recorded by two-photon calcium imaging in awake mice in vivo. The dataset includes several preprocessed features such as direction selectivity, frequency selectivity and natural scene selectivity. For the RORb cre-line, there are 40 experiments in total.

The visual cortex has been anatomically classified into seven subregions, six of which are featured in the Allen Brain Atlas: primary (VISp), anterolateral (VISal), anteromedial (VISam), lateral (VISl), posteromedial (VISpm), and rostrolateral (VISrl) [2,3,4]. It has been shown that certain neuronal subtypes may exhibit differences in selectivity and frequency response across visual subregions [4].

Allen Mouse Brain Connectivity Atlas provides a projection map from brain-wide area under the cre-independ or cre-dependent condition. Projection experiments are conducted using the recombinant AAV (rAAV). In visual area specific experiments, the areas are aligned with the visual areas in Allen Brain Observatory that it uses ISI to define at least six visual regions: VISp, VISal, VISrl, VISam, VISl, VISpm. rAAV is injected at the specific site and the site is recorded using serial 2D tomography. The data is then processed for the future analysis. The preprocessed data including the injection site and projection site and their voxel, projection volume (i.e. size of the target compared to injection site), projection density (i.e. size of the target compare to the projection site) and projection power (i.e. intensity of the pixel). 12 experiments for RORb cre-line are performed in one of the visual areas mentioned above. 


**Hypothesis**

We hypothesize that by combining datasets across electrophysiology, visual response, and connectivity, we may identify novel subtypes of RORb+ neurons with distinct physical properties, visual selectivity, and/or anatomical targets.

Below are useful methods for computing statistical tests for future use. We assess significance only after visualizing the data distribution.

```python
# Defining Useful Functions

def anova(factor, dataframe):
    v = []
    for i in dataframe["region"].unique():
        row = dataframe[dataframe["region"] == i]
        v.append(row[factor].dropna())
    s, p = sp.stats.f_oneway(v[0], v[1], v[2], v[3], v[4], v[5])
    return p

def spearman(fac1, fac2, dataframe):
    sub = dataframe[[fac1, fac2]].dropna()
    s, p = sp.stats.spearmanr(sub[fac1], sub[fac2])
    return s,p

def two_stats_ephys(variable, factor, cat1, cat2, dataframe, alpha):
    vals_1 = dataframe[dataframe[factor] == cat1]
    vals_2 = dataframe[dataframe[factor] == cat2]
    
    vals_1 = vals_1[variable].to_numpy()
    vals_2 = vals_2[variable].to_numpy()
    
    vals_1 = vals_1 [np.logical_not(np.isnan(vals_1))]
    vals_2 = vals_2 [np.logical_not(np.isnan(vals_2))]
    
    #test skew
    m1, s1 = stats.skewtest(vals_1)
    m2, s2 = stats.skewtest(vals_2)
    
    #apply statistical test
    if (s1 < alpha or s2 < alpha):
        stat, p = stats.mannwhitneyu(vals_1, vals_2)
        cd = (np.nanmedian(vals_1) - np.nanmedian(vals_2))
        my_test = "Mann-Whitney-U"
    else:
        stat, p = stats.ttest_ind(vals_1, vals_2)
        cd = (np.nanmean(vals_1) - np.nanmean(vals_2))
        my_test = "Independent-T"
            
    return my_test, p, s1, s2
```

**Allen Cell Types Database** 

We first sought to characterize the physical properties of RORb+ neurons in relation to all cells in the Cell Types Database. 

```python
#Get and format metadata, dropping metrics that are not of use to us
cell_df = pd.DataFrame(ctc.get_cells(species=[CellTypesApi.MOUSE])).set_index('id')
cell_df.drop(['cell_soma_location','apical', 'name','disease_state','donor_id','reconstruction_type','reporter_status', 'species', 'structure_area_id','structure_hemisphere'], axis=1, inplace=True)

#Get and format ephys data, dropping metrics that are not of use to us
ephys_df = pd.DataFrame(ctc.get_ephys_features()).set_index('specimen_id')
ephys_df.drop(['has_burst', 'has_delay', 'has_pause', 'id', 'seal_gohm', 'electrode_0_pa', 'thumbnail_sweep_id', 'rheobase_sweep_id', 'rheobase_sweep_number'], axis=1, inplace=True)

#Merge
cell_ephys_df = cell_df.join(ephys_df)
```

```python
# Compute various action potential widths for each stimulus and for each major point in the waveform
for stim in ["long_square", "ramp", "short_square"]:
    for meas in ["fast_trough", "peak", "slow_trough", "trough"]:
        cell_ephys_df["time_to_" + meas + "_" + stim] = cell_ephys_df[meas + "_t_" + stim] - cell_ephys_df["threshold_t_" + stim]

# Compute various action potential heights for each stimulus and for each major point in the waveform
for stim in ["long_square", "ramp", "short_square"]:
    for meas in ["fast_trough", "peak", "slow_trough", "trough"]:
        cell_ephys_df["volts_to_" + meas + "_" + stim] = cell_ephys_df[meas + "_v_" + stim] - cell_ephys_df["threshold_v_" + stim]        

# Compute minimum depolarization voltage for each stimulus type

for stim in ["long_square", "ramp", "short_square"]:
    cell_ephys_df["depol_" + stim] = cell_ephys_df["threshold_v_" + stim] - cell_ephys_df["vrest"]

    
#drop all metrics that represent raw time stamps and voltages (except threshold voltage and vrest)
for stim in ["long_square", "ramp", "short_square"]:
    for meas in ["fast_trough", "peak", "slow_trough", "trough"]:
        cell_ephys_df.drop([meas + "_t_" + stim], axis = 1, inplace = True)
        cell_ephys_df.drop([meas + "_v_" + stim], axis = 1, inplace = True)
    cell_ephys_df.drop(["threshold_t_" + stim], axis = 1, inplace = True)
```

```python
#creating dataframe with our cell types of interest
visp_df = cell_ephys_df[cell_ephys_df["structure_area_abbrev"] == "VISp"]
visp_df["by"] = "VISp neurons"
rorb_df = visp_df[visp_df["transgenic_line"] == "Rorb-IRES2-Cre"]
rorb_df["by"] = "VISp RORb+ neurons"
total_ephys = pd.concat([visp_df, rorb_df])

total_ephys.head()
```

### Figure 1 - Single Action Potential Properties of RORb+ Neurons

In order to characterize the electrophysiological properties of RORb+ Neurons, we compare select single action potential properties between RORb+ VISp neurons and the set of all VISp neurons in the Allen Cell Types Atlas. 

**A.** RORb+ neurons have a slightly higher AP width (computed as trough_time - threshold_time) than the set of all VISp neurons. 

**B.** RORb+ neurons have a higher depolarization voltage (computed as threshold_voltage - RMP) than the set of all VISp neurons.

Since, in each comparison, at least one distribution being compared was skewed, we used the Mann-Whitney-U test to assess the significance of the differences between each pair of distributions. The difference between AP width was significant (p = 1.0e-5), and the difference between depolarization voltage was also significant (p = 1.1e-6)

Collectively, the data suggests that, with a wider AP waveform and a larger stimulus required to trigger RORb+ neurons as compared to the entire VISp cell set, RORb+ have a tendency towards lower excitability. 

```python
fig, axs = plt.subplots(1,2, figsize = (15,5)) 

fig.suptitle("Fig 1. Single AP Properties of RORB+ Neurons compared to Cell Types Dataset")

sns.violinplot(ax = axs[0], x = total_ephys["by"],  y = total_ephys["time_to_fast_trough_long_square"] )
axs[0].set_ylabel("Time to Trough (s)")
axs[0].set_title("Action Potential Width")
t, p, s1, s2 = two_stats_ephys("time_to_fast_trough_long_square", "by", "VISp RORb+ neurons", "VISp neurons", total_ephys, 0.01)
axs[0].set_xlabel("\n" + t + " : p = " + str(p) + "\n Skew (RORb) : p = " + str(s1)+ "\n Skew (VISp) : p = " + str(s2))

sns.violinplot(ax = axs[1], x = total_ephys["by"],  y = total_ephys["depol_long_square"] )
axs[1].set_ylabel("Voltage (mV)")
axs[1].set_title("Depolarization Voltage")
t, p, s1, s2 = two_stats_ephys("depol_long_square", "by", "VISp RORb+ neurons", "VISp neurons", total_ephys, 0.01)
axs[1].set_xlabel("\n" + t + " : p = " + str(p) + "\n Skew (RORb) : p = " + str(s1)+ "\n Skew (VISp) : p = " + str(s2))

plt.show()
```

### Figure 2 - Spike Train Properties of RORb+ Neurons

We then plotted metrics related to spike trains generated by RORb+ neurons.  

**A.** RORb+ neurons have a higher latency compared to the set of all VISp neurons. 

**B.** RORb+ neurons have a higher average interspike interval (ISI) than the set of all VISp neurons.

**C** RORb+ neurons have a lower FI curve slope than the set of all VISp neurons, suggesting that more current is needed to produce a similar frequency spike train to VISp neurons.

Since, in each comparison, at least one distribution being compared was skewed, we used the Mann-Whitney-U test to assess the significance of the differences between each pair of distributions. Collectively, the data suggests that RORb+ neurons take longer to produce their first spike (latency), have larger intervals between spikes (avg ISI) and require a larger current to produce comparable frequency spike trains (FI curve slope). ### Figure 3 - Principal Component Analysis of VISp Neurons by Electrophysiological Properties

After comparing VISp RORb+ neurons to the set of VISp neurons, we wanted to determine if RORb+ neurons could be discerned by PCA analysis of electrophysiologial properties, and if subtypes existed within RORb+ neurons. 

**A.** Principal component analysis of VISp neurons. Each point is colored by dendritic type. 

**B.** Principal component analysis of VISp neurons. Points are colored by whether or not they are RORb+.

Panel A reveals a weak separation of inhibitory versus excitatory neurons (described by dendritic type), across PCA 2. Panel B shows that most RORb+ neurons are present in the segment of the graph containing excitatory neurons, confirming that, based on the metrics in the Allen Cell Types Database, RORb+ neurons are similar to other RORb- excitatory cells. However, RORb+ neurons failed to cluster distinctly and separately from other cell types, suggesting that these neurons do not form a distinct subset of neurons based solely on electrophysiological response. Furthermore, we failed to identify clean subclusters of RORb+ neurons, suggesting that electrophysiological properties may not be sufficient to identify or describe heterogeneity within RORb+ neurons. 

```python
#Drop all rows with NA
pca_df = cell_ephys_df.dropna()

#Normalize
pca_norm = pca_df.drop(columns=['structure_layer_name','structure_area_abbrev','transgenic_line','dendrite_type'])
pca_norm = (pca_norm - pca_norm.mean())/pca_norm.std() 
```

### Figure 7 - Correlation Analysis between Selectivity Indices

Given the similarity in trends between OSI and DSI, as well as among TFDI, SFDI, Image Selectivity in RORb+ neurons across brain regions, we asked the question whether correlations existed between these metrics, in order determine whether distinct subsets of neurons were responsible for exerting different types of selectivity, or whether subsets of neurons each drove a certain tendency towards selectivity.

Correlation between OSI and DSI was visualized in **A.** , and correlation among TFDI, SFDI, and Image Selectivity was visualized in **B-D.** .

**A.** OSI and DSI seem to have a weak positive correlation, with a higher density of neurons clustering at OSI = DSI = 1. **B.** SFDI and TFDI demonstrate weak positive covariance. **C.** SFDI and Image Selectivity demonstrate weak positive covariance. **D.** TFDI and Image Selectivity demonstrate extremely weak positive covariance, if any.

In order to assess whether each pair of variables displayed similar trends in monotonicity as well as determine strength of correlation, we used the Spearman Rank Correlation Test. By this test, OSI and DSI demonstrated weak correlation (R = 0.17, p = 6.6e-23). SFDI and TFDI demonstrated weak correlation (R = 0.30, p = 5.0e-67), SFDI and image selectivity demonstrated weak correlation (R = 0.31, p = 5.0e-67), and TFDI and image selectivity demonstrated weak correlation (R = 0.17, p = 5.0e-67).

Collectively, the weak correlations between OSI and DSI as well as among TFDI, SFDI, and image selectively suggest that different populations of neurons govern the tendency towards each of these behaviors, as opposed to distinct subtypes with high selectivity and subtypes with low selectivity. However, this purely correlative assesment requires direct experimental validation. Furthermore, the weak clustering of neurons in Figure A warrants further investigation, as to whether this cluster represents a population of neurons that drive both orientation and direction selectivity. 

```python
fig, axs = plt.subplots() 
fig.suptitle(" Figure 7. Correlations between Selectivity Indices of RORb+ neurons across All Visual Subregions")

axs.scatter(x = total["osi_dg"],  y = total["dsi_dg"] , alpha=0.1)
s, p = spearman("osi_dg", "dsi_dg", total)
axs.set_xlabel("OSI \n\n" + 'Spearman R - ' + str(s) + "\n p = " + str(p))
axs.set_ylabel("DSI")
axs.set_title("A. OSI vs DSI")


fig, axs = plt.subplots(1,3, figsize = (15,5)) 
axs[0].scatter(x = total["sfdi_sg"],  y = total["tfdi_dg"] , alpha=0.1)
s, p = spearman("sfdi_sg", "tfdi_dg", total)
axs[0].set_xlabel("SFDI \n\n" + 'Spearman R - ' + str(s) + "\n p = " + str(p))
axs[0].set_ylabel("TFDI")
axs[0].set_title("B. SFDI vs TFDI")

axs[1].scatter(x = total["sfdi_sg"],  y = total["image_sel_ns"] , alpha=0.1)
s, p = spearman("sfdi_sg", "image_sel_ns", total)
axs[1].set_xlabel("SFDI \n\n" + 'Spearman R - ' + str(s) + "\n p = " + str(p))
axs[1].set_ylabel("Image Selectivity")
axs[1].set_title("C. SFDI vs Natural Scene Selectivity")

axs[2].scatter(x = total["tfdi_dg"],  y = total["image_sel_ns"] , alpha=0.1)
s, p = spearman("tfdi_dg", "image_sel_ns", total)
axs[2].set_xlabel("TFDI \n\n" + 'Spearman R - ' + str(s) + "\n p = " + str(p))
axs[2].set_ylabel("Image Selectivity")
axs[2].set_title("D. TFDI vs Natural Scene Selectivity")

plt.show()
```

### Figure 8:  RORb+ projection density in other areas from visual cortex

There are different recording from the same regions. We can look at the mean projection density from all experiments to other brain regions. Projection density is defined as:<br>
projection density = sum of detected pixels / sum of all pixels in division (<a href = "http://help.brain-map.org/display/mouseconnectivity/API#API-ProjectionDataGridding">documentation</a>)<br>

In the graphs below, we use Kruskal-Wallis H test to see whether the tops are significantly different or not until we find a group contain maximum number of structures with no significant difference. Kruskal-Wallis H test is a non-parametric statistical test to see whether there are at least one group is from a different distribution among two or more groups. It can be seen as a multi-comparison version of Mann-Whitney U Test. Here, we choose to use Kruskal-Wallis H test instead of ANOVA because the data is skewed.


A. Boxplot of the projection density in log scale. B. Barchart of the projection density with error bar representing standard deviation. We can see that besides visual areas, RORb+ cell projects to retrosplenial area, posterior auditory area, anterior cingulate area, nucleus raphe pallidus, caudoputamen, postsubiculum and temporal association area. Kruskal-Wallis p value is 0.069 for these groups of structures.

```python
name_map = structure_tree.get_name_map()

unionizes_mean = (unionizes.groupby('structure_id', as_index = False)['projection_density'].mean())
names = [name_map[strid] for strid in unionizes_mean['structure_id']]

unionizes_std = (unionizes.groupby('structure_id', as_index = False)['projection_density'].std())
set_mean_std = pd.DataFrame({'structure_name': names, 
                            'mean_projection_density': unionizes_mean['projection_density'], 
                            'stdev':  unionizes_std['projection_density']
                            })
set_mean_std.sort_values(by = 'mean_projection_density', ascending = False, inplace = True)

top20 = set_mean_std[:20]['structure_name']
names = [name_map[strid] for strid in unionizes['structure_id']]
unionizes['structure_name'] = names
pltdat = unionizes[unionizes['structure_name'].isin(top20)]
pltdat['structure_name_cat'] = pd.Categorical(pltdat['structure_name'], categories = list(top20), ordered = True)
pltdat = pltdat.sort_values('structure_name_cat')

projection_density_list = []
gk = pltdat.groupby('structure_name')
for i, item in top20.iteritems():
    projection_density_list.append(list(gk.get_group(item)['projection_density']))

idx = 0
for i in range(19):
    stat, p = stats.kruskal(*(projection_density_list)[:20-i])
    if p >= 0.05:
        idx = 20 - i
        break

list_to_drop = list(top20)[idx:20]
for item, i in zip(pltdat['structure_name_cat'], pltdat.index):
    if item in list_to_drop:
        pltdat = pltdat.drop(i)
                
fig, ax = plt.subplots(2, 1, figsize = (15, 20))
fig.subplots_adjust(hspace=0.7)
sns.boxplot(x="structure_name", y="projection_density", data = pltdat, order = list(top20[:idx]), ax = ax[0], palette = "Blues_r")
sns.swarmplot(x="structure_name", y="projection_density", data = pltdat, order = list(top20[:idx]), ax = ax[0], color=".25")
ax[0].set_yscale("log")
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation = 90)
ax[0].set_xlabel("Structure: Kruskal-Wallis p = %f" % p, fontsize = 18)
ax[0].set_ylabel('Projection Density in log', fontsize = 18)
ax[0].set_title('A. Boxplot of Projection Density', fontsize = 20)
sns.barplot(x="structure_name", y="projection_density", data = pltdat, ax = ax[1], ci = 'sd', palette = "Blues_r")
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation = 90)
ax[1].set_xlabel("Structure: Kruskal-Wallis p = %f" % p, fontsize = 18)
ax[1].set_ylabel('Projection Density', fontsize = 18)
ax[1].set_title('B. Barchart of Projection Density', fontsize = 20)
plt.suptitle('Fig 8. Top Structure Projection Density from Visual Cortex in RORb Neurons', fontsize = 22)
plt.show()
```

## Conclusion

**Discussion**

From the electrophysiology data, we first characterized the behavior of RORb neurons by comparing various ephys metrics between RORb neurons and the entire allen brain data set. In order to characterize the electrophysiological properties of RORb+ Neurons, we compared select single action potential properties between RORb+ VISp neurons and the set of all VISp neurons in the Allen Cell Types Atlas. We found that with a wider AP waveform and a larger stimulus required to trigger RORb+ neurons as compared to the entire VISp cell set (Figure 1) and that RORb neurons take longer to produce their first spike, have larger intervals between spikes, and require a larger current to produce comparable frequency spike trains (Figure 2). These results suggest that RORb+ neurons have a tendency towards lower excitability than the total VISp population in the cell types database. 

Then, we performed PCA on spiny RORb neurons to determine if RORb+ neurons could be discerned by PCA analysis of electrophysiological properties, and if subtypes existed within RORb+ neurons. Our clustering of PCA shows that a weak separation of inhibitory to excitatory neurons across PCA2 and most RORb+ neurons are present in the segment of the graph containing excitatory neurons (Figure 3), suggesting that RORb+ neurons do not form a distinct subset of neurons and have the similar properties to other RORb- excitatory cells. 

A major limitation of clustering based on PCA is that we failed to identify and clean subclusters of RORb+ neurons, suggesting that electrophysiological properties may not be sufficient to identify or describe heterogeneity within RORb+ neurons. This might be due to sampling biases. Additionally, since recording was done in vitro, electrophysiological properties might not reflect in vivo conditions, due to potential alterations in physical properties, gene expression, or lack of physiological connections.




We then wanted to ask the question whether subsets of RORb+ neurons displayed distinct properties across visual subregions. However, due to sparsity of data in the Allen Cell Types database for neurons not in VISp, we turned to the Allen Brain Observatory dataset. In order to comprehensively characterize neuronal response to visual coding, we looked at five different selectivity/discriminability indices, and found similarities in trends between OSI and DSI (Figure 4), as well as among TFDI, SFDI, and Image Selectivity (Figure 5, 6). However, we failed to find strong correlations between the former two indices and among the latter three indices (Figure 7), suggesting that neurons might not be primed to drive multiple types of selectivity by a common mechanism within the subset of RORb+ neurons. Rather, selectivity might be driven independently in different sets of neurons, or may be a consequence of circuit-level activity or anatomical connectivity. This hypothesis is further supported by the fact that neurons with certain selectivities are enriched or depleted in different subregions, suggesting context may be important in governing selectivity. 

A major limitation to the Allen Brain Observatory dataset is the fact that cell spiking activity is determined by calcium imaging rather than electrophysiology, due to technical limitations. While this serves as a good approximation, it is not a direct correlate and differs from electrical activity in kinetics. 

An interesting comparison to make with this data is to make a comparison between RORb+ neurons with other cell types in the Allen Brain Observatory, in order to get an idea of the relative selectivities of RORb+ neurons. Additionally, genetic studies and local connectivity studies may be useful in determining the neural basis for selectivity in RORb+ neurons. 



From the result we get from Allen Mouse Brain Connectivity dataset, we can conclude that there is a specific subtype of RORb+ positive neurons in the visual area, especially lateral visual cortex, that project to the areas outside the visual cortex such as posterior auditory cortex and retrosplenial area(figure 8, 10). While posterior auditory area is related with sound localization[11], retrosplenial area is associated with spatial cognition[12]. From the result, however, we cannot conclude what is the role of RORb+ in the connectivity between visual areas and these two non-visual areas. 

A possible explanation is that RORb+ probably served to axonal guidance between these regions, since there is evidence that RORb regulates the innevation in midbrain and barrel cortex development[8, 9]. Another possibility is that RORb+ regulates the neuronal subtype specification across these cortical areas, as evidence showing that it contributes to the subtype specification in neocortex layers [10].

The limitation of this project is not only due to the relatively small sample size, but also the actual study focus on the projection of RORb+ neurons. According to the text-mining result in PubMed, currently limited numbers of research are conducted on the RORb+ function in visual cortex and other brain areas. In the aspect of data science, RORb expression level across different ages and its relation with other neocortex specific genes are potential future analysis to see whether RORb is crucial for early visual cortex development and innervation to other brain areas. Based on the maximum projection images of the experiment (figure 9), the projection pattern difference between ipsilateral and contralateral of the brain may exist, and further investigation of the relation between RORb+ neurons and the connectivity of two hemispheres is promising.
