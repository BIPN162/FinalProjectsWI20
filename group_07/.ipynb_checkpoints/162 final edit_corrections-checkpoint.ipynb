# Group 7: Analyzing the firing activity of neurons within the Hippocampus

## Overview
	This study utilizes the Connectivity and Neuropixel datasets from the Allen Research Institute to examine possible links between connectivity and functional correlation in the hippocampus.  We showed that strong local connection between CA3 neurons predicted stronger correlation of local neuron activity, compared to correlation with CA2 neurons which received weaker projection from CA3. These results lead to better understanding of the possible mechanisms where connected regions can influence each other during perception. 

## Research Question:
    Does connectivity influence function: do cells in connected brain regions correlate more to each other in terms of their firing activity?

## Background & Prior Work
    The datasets used for this project were collected by the Allen Research Institute.The Neuropixel dataset contains data taken from neuropixel probes, which records the electrical activity occurring in the structure(s) of interest with single unit resolution and large channel counts. This is made possible by a specialized coating that dramatically lowers the capacitance of the probe allowing large number of channels to be compacted into a relatively small format without considerable increase of noise.  The probes are coated with fluorescent dye that allows the insertion track to be visualized after recording.  An image processing software is used to assess the amount of tissue stretch around the probe trajectory and assign the location of each channel according to which brain region it landed in (1).  Allen’s Connectivity dataset utilizes virus injection to track axons of projection neurons as they travel to different downstream regions. A recombinant adeno-associated virus (rAAV) is injected in experimental animals that express Cre recombinase in specific cell types to allow for the expression of fluorescent dye in selected populations. The transgenic Cre mouse will express the virus in the tagged cell types, as the Cre recombinase allows expression of the fluorescent protein, or other protein of interest, contained in a double-floxed inverted orientation or (DIO) region. Furthermore, rAAV tracers can also be designed to carry proteins that tag anterograde synapses, allowing for the tracing of distinct neural pathways from the structures of interest.(2)    
    
    Combination of the Neuropixel and Connectivity datasets allows us to investigate the link between connectivity and function of structures that appear in both datasets. This includes the CA1, CA2, and CA3 regions of the hippocampus, and the anteromedial visual cortex. Integration of these datasets allows us to visualize not only the projection density and neural networks in these regions (connectivity data), but also the individual activity of neurons within those networks during certain tasks (neuropixel data). This project focuses on the projections and activity within the hippocampal region.
        
    The CA3 region consists of strong connections with other brain regions, as well as with itself. These autoassociative networks are key in memory formation and retrieval through pattern completion. Pattern completion is a distinctive trait of the CA3 autoassociative network whereby an incomplete cue that leads to the firing of a few neurons in a memory engram can go on to activate the rest of the engram and retrieve the memory (3,4). The spiking rate and timing plays a vital role in the robustness of this memory event(3). CA3 also projects to the CA1 region of the hippocampus. These connections are known as the Schaffer collaterals. It has been observed that the Schaffer collaterals have more clustering of synaptic activity, possibly indicating convergence of CA3 neurons onto CA1 neurons. Furthermore, the majority of neurons projecting from CA3 to CA1 come from the distal region of CA3(4). In short, the CA3 and CA1 regions of the hippocampus have complex networks where the specifics of population activity impacts the functionality in memory. Studying their behavior sheds insight on the physiological mechanisms behind memory.

## Hypothesis
    Brain regions that are more connected (with higher projection density) should have higher correlation between neurons in each region.

## Finding overlap between Connectivity Data and Neuropixel Data

```python
# get experiments where the viral injection landed in one of the areas specified above
overlap = mcc.get_experiments(dataframe=True,injection_structure_ids = ids)

# for loop to limit search only to experiments where most of the virus is in the desired structures
# disregard experiments where the structures were infected by accidental spread
in_str_list = []
for id, exp in overlap.iterrows():
    primary_injection_structure=exp.primary_injection_structure
    #Append true if primary injection structure is in our list of structure ids
    in_str_list.append(primary_injection_structure in ids)

# DataFrame of experiments on structures that appear in both datasets
overlap = overlap[in_str_list]
```

```python
# create a digraph to store projections
str_name = overlap.structure_name.unique()  #get list of structures found in the overlap
stct_ids = overlap.structure_id.unique()    #get ids of structures found in the overlap
label={}                                    #create a dictionary of names for plotting algorithm
attr_lst=['id', 'acronym', '_name']         #list of attributes each node would have
Vctx_cnctn = nx.DiGraph()                   #create a graph with directed edge using networkx
for id in stct_ids:
    name = G.nodes[id]['_name']     #get structure fullname corresponding to id
    Vctx_cnctn.add_node(name)       #add node to the directed graph with the structure fullname as its name attribute
    label[name] = name              #add the name to dictionary of names for plotting algorithm
    for attr in attr_lst:           #append id, acronym, name attributes to the created node
         Vctx_cnctn.nodes[name][attr] = G.nodes[id][attr]
```

```python
# Model all recorded projections as DiGraph in networkx
print("%d total experiments" % len(overlap))
i = 0 
for id, exp in overlap.iterrows():
    id = [id]                                #id of the experiment, a list is required for the allensdk method
    cre = exp.transgenic_line                #store Cre line of this experiment
    spread = len(exp.injection_structures)   #store the number of structures that the virus spread to
    if cre==None:
        cre = 'wildtype'
    # each node will be a population defined by structure and cell type
    str_id = exp.structure_id                #store the structure id of this experiment
    str_nm = G.nodes[str_id]['_name']        #get name of structure using ontology graph
    nd_nm = str_nm+'  '+cre                  #set node name as the structure name and cre name
    
    #get Allen's characterization of projections from this experiment
    exp_uni = mcc.get_structure_unionizes(experiment_ids = id)
    
     #remove projections to structures that are not included in Neuropixel data
    target = exp_uni[exp_uni['structure_id'].apply(lambda x: x in ids)] 
    
    #remove projections to structures that have 0 projection density
    target = exp_uni[exp_uni['projection_density'].apply(lambda x:x!=0)]

    # looking at projection density across both hemispheres
    target =  target[target['hemisphere_id']==3]
    
    # focus on fluorescence that is not at the injection site
    target_not_injection = target[~target['is_injection']]
    
    for _ , projection in target_not_injection.iterrows():
        
        # set projection strength (edge weight) to the recorded projection density divided by spread
        prjctn_str = projection.projection_density /spread
        
        # set end_id to the structure where fluorescence is detected, as long as the structure is in the specified list above
        end_id = projection.structure_id 
        if not end_id in ids:
            continue
        
        # store the name of the end structure
        end_str =  G.nodes[end_id]['_name']
        
        #create a new node for the structure and cell type defined population
        Vctx_cnctn.add_node(nd_nm)
        
        #the new node is rooted if there is an edge pointing from its parent structure to it
        rooted = Vctx_cnctn.has_edge(str_nm,nd_nm)
        # check if a projection from the population to the target structure already exists
        connected = Vctx_cnctn.has_edge(nd_nm,end_str)
        #create an edge from the parent structure to the population to root the node
        if not rooted: 
            Vctx_cnctn.add_edge(str_nm,nd_nm)
         #create the projection to the new node if it does not yet exist
        if not connected: 
            Vctx_cnctn.add_edge(nd_nm,end_str,weight = prjctn_str)
            Vctx_cnctn[nd_nm][end_str]['sum'] = 1               #create attribute sum,which is number of total records of this projection
            Vctx_cnctn[nd_nm][end_str]['sum_w'] = prjctn_str    #create attribute sum of weights (projection densities)
            Vctx_cnctn[nd_nm][end_str]['weight'] = prjctn_str   #assign weight to weight attribute
        # if the projection already exists:
        else:
            sum_weight = Vctx_cnctn[nd_nm][end_str]['sum_w']               
            sum_weight += prjctn_str
            Vctx_cnctn[nd_nm][end_str]['sum_w'] = sum_weight    #add the new weight to sum weight
            Vctx_cnctn[nd_nm][end_str]['sum'] += 1              #increase sum attribute by one
            pjt_sum = Vctx_cnctn[nd_nm][end_str]['sum']
            Vctx_cnctn[nd_nm][end_str]['weight'] = sum_weight/pjt_sum     # recalculate average weight from sum and sum weight
    i+=1                                                                  #increment counter
    # report progress every for every 5 experiments analyzed 
    if i%5 == 0:
        print(f'{i} experiments analyzed')
```

```python
# helper function for turning spike time data into firing rate
def get_st(lst):
    # input is a list of unit ids
    st = []
    # append the spike times of unit ids in lst to st
    for id in lst:
        st.append(session.spike_times[id])
    return st
```

```python
# helper function for turning spike time data into firing rate
def bin_st(bin_size, start, end, st):
    #bin_size: the size of time bin to compute firing rate
    #start: start time of trial
    #end: end time of trial
    #st: spike time
    
    #duration of trial
    duration = end - start
   
    #number of units in spike time list
    numOfUnits = len(st)
    
    #number of bins for this trial
    numOfBins = int(round(duration/bin_size))
    
    #x is the container for binned firerate
    x = np.zeros((numOfUnits,numOfBins))
    
    #increase spike count of a time bin by one if a spike time lands in a particular time bin
    for i in range(len(st)):
        tmp = st[i][(st[i]>=start) & (st[i]<=end)]
        for j in range(len(tmp)):
            bin_id = int(round((tmp[j]-start) /bin_size)-1)
            x[i][bin_id]+=1
    return x
```

```python
# helper function to find Spearman correlation distribution between rows of two matrices with uneven number of rows and columns

#ans = table of r values(M*N) M and N are number of rows(neurons) in the two matrices
#is_sig = table of boolean (M*N) of whether the correlation is significant (True and False)
def find_corr(x,y):
    
    #check if x and y are the same matrix
    same_area = np.array_equal(x,y)
    
    #containers for row correlation matrix and significance matrix
    ans = np.zeros((len(x),len(y)))
    is_sig = np.zeros((len(x),len(y)))

    for i in range(len(x)):
        
        # disregard correlation between same rows if x and y are the same matrix
        y_ = y.copy()
        if same_area:
            y_ = np.delete(y,i,0)
        #assign rows from matrix 1 and 2 to xi and yj
        xi = x[i]
        for j in range(len(y_)):
            yj = y_[j]
            
            #store correlation bewteen ith row and jth column in x and y in the output matrix
            ans[i][j] = stats.spearmanr(xi,yj)[0]
            p = stats.spearmanr(xi,yj)[1]
            # store significance boolean in significance matrix
            is_sig[i][j] = p < 0.05
    return ans,is_sig       
```

```python
# helper function to round to appropriate significant figures
from math import log10, floor
def round_sig(x, sig=2):
    return round(x, sig-int(floor(log10(abs(x))))-1)
```

```python
# bin the spike time data and average firing rate across trials
bin_size = 1

# number of trials with natural_movie_three
num_of_tr = len(start)

#prepare two np zero arrays in the correct size (rows = number of neurons in region, columns = length of movie / bin size)
CA1_fr = np.zeros((len(CA1_ids),round(600.50163/bin_size)))
CA3_fr = np.zeros((len(CA3_ids),round(600.50163/bin_size)))

# bin the spike time data from each trial for each region and input to the np array
for i in range(num_of_tr):
    strt = start[i]
    nd = end[i]
    CA1_fr+=bin_st(bin_size,strt,nd,CA1_st)
    CA3_fr+=bin_st(bin_size,strt,nd,CA3_st)

# average the firing rate across trial, remove all cells that did not fire
CA1_fr = CA1_fr/num_of_tr
CA3_fr = CA3_fr/num_of_tr
CA1_fr= CA1_fr[~np.all(CA1_fr == 0,axis=1)]
CA3_fr= CA3_fr[~np.all(CA3_fr == 0,axis=1)]

# average firing rates across neurons in CA1 and CA3
CA1_avg= np.average(CA1_fr,axis = 0)
CA3_avg= np.average(CA3_fr,axis = 0)

# average the firing rate across time for CA1 and CA3 cells
CA1_bin_avg = np.average(CA1_avg)
CA3_bin_avg = np.average(CA3_avg)
```

```python
# calculate the p values for this difference using a student's t test
z,p = stats.ttest_rel(per_cell_avg_corr_CA3_CA1,per_cell_avg_corr_CA3_CA3)
print(f'the p value for this difference is {p}')
```

A paired t test is performed for this analysis (n=27) since both groups of correlation coefficients are averaged for each CA3 neuron, and the identity of these cells can be traced across the two groups.  With the limited data in this session, we see a trend where for each neuron in CA3, its average correlation coefficients with neurons the local CA3 population is slightly higher than its average correlation coefficients with cells in CA1.  Even though the trend is coherent with our hypothesis that strong projection between two population should indicate more correlated cell activity, the local projection density for CA3 is only slightly higher than its projection density to CA1 and the difference between the two distributions is not significant.  Let's take a look at the distribution of correlation coefficients for each cell in CA3 with all cells in the downstream areas.

```python
# what should we expect the distribution of correlation coefficients to be?
# let's examine the correlation between random independent Poisson neurons with the same average firing rate as neurons in the CA regions

# generate a random firing rate for one fake 'CA3' upstream neuron
num_of_bin=round(600.50163/bin_size)
rand_CA3=np.zeros((1,num_of_bin))
rand_CA3[0] = np.random.poisson(lam = CA3_bin_avg,size = num_of_bin)

# generate a random firing rate for independent 'downstream neurons'
# function to generate the firing pattern of n random neurons from a mean firing rate 
def rand_n_neurons(average,n_neurons,num_of_bin):
   
    # matrix to hold random firing data
    rand_matrix = np.zeros((n_neurons,num_of_bin))
    
    # generate random firing pattern for each neuron
    for i in range(n_neurons):
        rand_matrix[i]=np.random.poisson(lam = average,size = num_of_bin)
    return rand_matrix

# simulate activity of random downstream populations of different sizes with the average firing rate of CA1 neurons as the mean
# naming scheme: rand_(structure name)_(number of neuron)
rand_CA1_100 = rand_n_neurons(CA1_bin_avg, 100, num_of_bin)
rand_CA1_500 = rand_n_neurons(CA1_bin_avg, 500, num_of_bin)
rand_CA1_1000 = rand_n_neurons(CA1_bin_avg, 1000, num_of_bin)
```

```python
# analyze_trial analyzes the correlation of firing rates between CA3 neurons and CA1, CA2 and CA3 neurons
#if there are enough cells for analysis (>10 cells)

#=============This is the same code as detailed above but put into one function================

#input:ids=id of session
#returns tuple of:(corr_CA3_CAX,sig_CA3_CAX,CA3_CAX_p,CA3_CAX_z)
#corr_CA3_CAX= list of correlation coefficients matrix of CA3 cells with CA1,CA2 and CA3 in that order
#sig_CA3_CAX= list of correlation coefficients significance matrix of CA3 cells with CA1,CA2 and CA3 in that order
#CA3_CAX_p=list of p value for Mann Whitney test for distribution of correlation coefficients of each cell in CA3 to cells in down stream regions(in order of CA1,CA2,CA3) compared to a random poisson population
#CA3_CAX_z=list of z value for Mann Whitney test described above
def analyze_trial(ids):
    #path to the session nwb file on disk
    nwb_path = f'C:\\allen data\\ecephys_session_{ids}.nwb'
    #read session data to session
    session = EcephysSession.from_nwb_path(nwb_path, api_kwargs={
            "amplitude_cutoff_maximum": np.inf,
            "presence_ratio_minimum": -np.inf,
            "isi_violations_maximum": np.inf
        })
    #get all units from this session with signal to noise ratio bigger than 2
    units = session.units[session.units['snr'] > 2]

    #get id of units in CA1,CA2,CA3 respectively
    CA1_ids = units[units['ecephys_structure_acronym']=='CA1'].index.values
    CA2_ids =units[units['ecephys_structure_acronym']=='CA2'].index.values
    CA3_ids =units[units['ecephys_structure_acronym']=='CA3'].index.values
    num_of_CA1_cells=len(CA1_ids)
    num_of_CA2_cells=len(CA2_ids)
    num_of_CA3_cells=len(CA3_ids)
    enough_CA3_cells=num_of_CA3_cells>10
    enough_CA2_cells=num_of_CA2_cells>10
    enough_CA1_cells=num_of_CA1_cells>10

    if not enough_CA3_cells:
        print(f'Trial {ids} do not have enough CA3 cells')
        return None
    else:
        #find epoch of natural movies
        #get all stimulus epochs of the session
        epochs = session.get_stimulus_epochs()
        #limit to only stimulus type: natural_movie_three, which is a 600 second movie
        epochs = epochs[epochs.stimulus_name =='natural_movie_three']
        #assign start and end time of all trials of natural_movie_three
        start = list(epochs.start_time)
        end = list(epochs.stop_time)
        #bin the spike time data and average firing rate across trials
        bin_size = 1
        num_of_bin=round(600.50163/bin_size)
        #number of trials with natural_movie_three
        num_of_tr=len(start)

        #---------------run CA3_CA3 analysis----------------
        
        #get spike time data for all CA3 cells
        CA3_st = get_st(CA3_ids)
        #bin and add up firing rate during trials of natural movie
        CA3_fr=np.zeros((num_of_CA3_cells,num_of_bin))
        for i in range(num_of_tr):
            strt = start[i]
            nd = end[i]
            CA3_fr+=bin_st(bin_size,strt,nd,CA3_st)
        #average firing rate accross trial
        CA3_fr = CA3_fr/num_of_tr
        #delete cells that didn't fire
        CA3_fr= CA3_fr[~np.all(CA3_fr == 0,axis=1)]
        #find average firing rate
        CA3_bin_avg = np.average(CA3_fr)
        #find distribution of correlation coefficients between CA3 and CA3 cells
        corr_CA3_CA3,sig_CA3_CA3=find_corr(CA3_fr,CA3_fr)
        #find distribution of correlation coefficients between random poisson cell with CA3 firing rate and random CA3 
        #cells with same popultaion size as recorded CA3 population
        randcorr_CA3_CA3,sig_CA3_CA3=find_corr(CA3_fr,rand_n_neurons(CA3_bin_avg,num_of_CA3_cells,num_of_bin))
        #statistics comparing distribution of firing correlation between random CA3_CA3 population and real CA3_CA3 population
        CA3_CA3_p_value =[stats.mannwhitneyu(corr_CA3_CA3[i],randcorr_CA3_CA3[i])[1] for i in range(num_of_CA3_cells)]
        CA3_CA3_z_score =[stats.mannwhitneyu(corr_CA3_CA3[i],randcorr_CA3_CA3[i])[0] for i in range(num_of_CA3_cells)]

        if enough_CA2_cells:
            #-------------------run CA3_CA2 analysis-----------------------
            #get spike time data for all CA2 cells
            CA2_st = get_st(CA2_ids)
            #bin and add up firing rate during trials of natural movie
            CA2_fr=np.zeros((num_of_CA2_cells,num_of_bin))
            for i in range(num_of_tr):
                strt = start[i]
                nd = end[i]
                CA2_fr+=bin_st(bin_size,strt,nd,CA2_st)
            #average firing rate accross trial
            CA2_fr = CA2_fr/num_of_tr
            #delete cells that didn't fire
            CA2_fr= CA2_fr[~np.all(CA2_fr == 0,axis=1)]
            #find average firing rate
            CA2_bin_avg = np.average(CA2_fr)
            #find distribution of correlation coefficients between CA3 and CA2 cells
            corr_CA3_CA2,sig_CA3_CA2=find_corr(CA3_fr,CA2_fr)
            #find distribution of correlation coefficients between random poisson cell with CA3 firing rate and random CA2 
            #cells with same popultaion size as recorded CA2 population
            randcorr_CA3_CA2,sig_CA3_CA2=find_corr(CA3_fr,rand_n_neurons(CA2_bin_avg,num_of_CA2_cells,num_of_bin))
            #statistics comparing distribution of firing correlation between random CA3_CA2 population and real CA3_CA2 population
            CA3_CA2_p_value =[stats.mannwhitneyu(corr_CA3_CA2[i],randcorr_CA3_CA2[i])[1] for i in range(num_of_CA3_cells)]
            CA3_CA2_z_score =[stats.mannwhitneyu(corr_CA3_CA2[i],randcorr_CA3_CA2[i])[0] for i in range(num_of_CA3_cells)]
        else:
            corr_CA3_CA2=[]
            sig_CA3_CA2=[]
            CA3_CA2_p_value=[]
            CA3_CA2_z_score=[]

        if enough_CA1_cells:
            #----------------run CA3_CA1 analysis----------------
            #complete code for analyzing CA3_CA1 correlation
            #get spike time data for all CA1 cells
            CA1_st = get_st(CA1_ids)
            #bin and add up firing rate during trials of natural movie
            CA1_fr=np.zeros((num_of_CA1_cells,num_of_bin))
            for i in range(num_of_tr):
                strt = start[i]
                nd = end[i]
                CA1_fr+=bin_st(bin_size,strt,nd,CA1_st)
            #average firing rate accross trial
            CA1_fr = CA1_fr/num_of_tr
            #delete cells that didn't fire
            CA1_fr= CA1_fr[~np.all(CA1_fr == 0,axis=1)]
            #find average firing rate
            CA1_bin_avg = np.average(CA1_fr)
            #find distribution of correlation coefficients between CA3 and CA1 cells
            corr_CA3_CA1,sig_CA3_CA1=find_corr(CA3_fr,CA1_fr)
            #find distribution of correlation coefficients between random poisson cell with CA3 firing rate and random CA1 
            #cells with same popultaion size as recorded CA1 population
            randcorr_CA3_CA1,sig_CA3_CA1=find_corr(CA3_fr,rand_n_neurons(CA1_bin_avg,num_of_CA1_cells,num_of_bin))
            #statistics comparing distribution of firing correlation between random CA3_CA1 population and real CA3_CA1 population
            CA3_CA1_p_value =[stats.mannwhitneyu(corr_CA3_CA1[i],randcorr_CA3_CA1[i])[1] for i in range(num_of_CA3_cells)]
            CA3_CA1_z_score =[stats.mannwhitneyu(corr_CA3_CA1[i],randcorr_CA3_CA1[i])[0] for i in range(num_of_CA3_cells)]
        else:
            corr_CA3_CA1=[]
            sig_CA3_CA1=[]
            CA3_CA1_p_value=[]
            CA3_CA1_z_score=[]
        #combine result of all caculation into master lists and return master lists
        corr_CA3_CAX=[corr_CA3_CA1,corr_CA3_CA2,corr_CA3_CA3]
        sig_CA3_CAX=[sig_CA3_CA1,sig_CA3_CA2,sig_CA3_CA3]
        CA3_CAX_p=[CA3_CA1_p_value,CA3_CA2_p_value,CA3_CA3_p_value]
        CA3_CAX_z=[CA3_CA1_z_score,CA3_CA2_z_score,CA3_CA3_z_score]
        return (corr_CA3_CAX,sig_CA3_CAX,CA3_CAX_p,CA3_CAX_z)
```

```python
#now we can analyze all sessions containing recordings from CA3,CA1 and CA2 using a parallel for loop
#define a main function to run the parallel processing
def main():
    #using the concurrent future package we can submit the function analyze_trial with the argument ids for each id in relevant_sessions
    with concurrent.futures.ProcessPoolExecutor(max_workers=5) as excutor:
        results=[excutor.submit(analyze_trial,ids) for ids in relevant_sessions]
        #report as analysis of sessions are completed
        i=0
        for f in concurrent.futures.as_completed(results):
            i+=1
            print(f'{i} session completed')
    # returns results
    return(results)
# 'guarding the function'  Apparently required for the concurrent futures module
if __name__ == '__main__':
    results=main()
    
#this is essentially the same as running
for ids in relevant_sessions:
    results=[analyze_trial(ids) for ids in relevant_sessions]

#unfortunately jupyter notebook don't work well with the parallel processing module, so the equavalent code of above was run in spyder
#two of the 15 sessions identified above: session 793224716, 799864342 have some problem that caused them unable to be analized by our code, potentially due to file damage during download
```

Even though the difference between the means of two groups is significant(p=0.0184), this still demands further analysis as we can see that the average r-value for each groups are all very small (less than 0.06)!

```python
#let's take a look at how the average correlation coefficient for every CA3 cell
#and its downstream structures scatter around their average

f,ax=plt.subplots()
per_cell_avg_corr_pd = pd.DataFrame(columns=['correlation coefficient','session','group'])
for i in range(len(per_cell_avg_corr_CA3_CA1)):
    row = pd.DataFrame({'correlation coefficient':per_cell_avg_corr_CA3_CA1[i]})
    row['session'] = f'session {i+1}'
    row['group']='CA3 correlated with CA1'
    per_cell_avg_corr_pd=pd.concat([per_cell_avg_corr_pd,row])
for i in range(len(per_cell_avg_corr_CA3_CA3)):
    row = pd.DataFrame({'correlation coefficient':per_cell_avg_corr_CA3_CA3[i]})
    row['session'] = f'session {i+1}'
    row['group']='CA3 correlated with CA3'
    per_cell_avg_corr_pd=pd.concat([per_cell_avg_corr_pd,row])
sns.boxplot(x="group", y="correlation coefficient",hue='session', palette='colorblind',data=per_cell_avg_corr_pd,ax=ax)

#create labels for each point
ax.set_title('average correlation of each CA3 cells \n with cells in downstream regions',fontsize='large')
```

Fig 10. Distribution of correlation coefficients for each CA3 cells will all cells in the downstream region across sessions.  We can see that the spread around the mean r value of each session is pretty substantial, and is much bigger than the difference between the mean.  We need to further the analysis and look at the distribution of correlation coefficients with downstream cells of each CA3 neuron.

```python
#let's compare how different the r value distributions of CA3 neurons with different downstream populations are, given
#that they have significantly different distributions of r values compared to random populations. The u value is the smaller 
#value among the number of times a randomly picked value from distribution A is bigger than distrubution B and vise versa
#This value should be around half the product between number of samples in A and number of sample in B.  The smaller the U
#value, the more difference between distribution A and B to interpreate diviation in shape from U scores, we should divid 
#the score by the total number of comparison.  Since we are comparing total of N correlation coefficients 
#(N being the size of downstream population) with N coefficients with random population of size N
#this value should be N^2

f,ax=plt.subplots()
#create dataframe to hold z scores
sig_z_pd = pd.DataFrame(columns=['normalized_U','session','group'])
#for each row(session) in CA3_CA1_z, save all z values to a new dataframe labeled with session and group if the corresponding p value for the z score is less than 0.05
i=0
for row in CA3_CA1_z:
    #get number of CA1 cells(N) from the shape of corr_CA3_CA1 of session i, this should be number of columns in this matrix
    _,num_of_CA1_cells=corr_CA3_CA1[i].shape
    #num of comparison should be N^2
    num_of_comparison=num_of_CA1_cells**2
    #append normalized U scores if the corresponding p value is less than 0.05
    row =[row[j]/num_of_comparison for j in range(len(row)) if CA3_CA1_p[i][j]<0.05]
    row = pd.DataFrame({'normalized U':row})
    row['session'] = f'session {i+1}'
    row['group']='CA3 compared to CA1'
    sig_z_pd=pd.concat([sig_z_pd,row])
    i+=1
i=0
#do the same thing with CA3_CA3_z
for row in CA3_CA3_z:
    #get number of CA1 cells(N) from the shape of corr_CA3_CA3 of session i, this should be number of columns in this matrix
    _,num_of_CA3_cells=corr_CA3_CA3[i].shape
    #num of comparison should be N^2
    num_of_comparison=num_of_CA3_cells**2
    #append normalized U scores if the corresponding p value is less than 0.05
    row =[row[j]/num_of_comparison for j in range(len(row)) if CA3_CA3_p[i][j]<0.05]
    row = pd.DataFrame({'normalized U':row})
    row['session'] = f'session {i+1}'
    row['group']='CA3 compared to CA3'
    sig_z_pd=pd.concat([sig_z_pd,row])
    i+=1
#plot grouped box plots 
sns.boxplot(x="group", y="normalized U",hue='session', palette='colorblind',data=sig_z_pd,ax=ax)

#create labels for each point
ax.set_title('Normalized Mann Whitney U of each CA3 cells with cells in downstream regions \n that have distribution of r values significantly different from random',fontsize='large')

#running a t-test to show that the distrubution of normaized U values are truly different
sig_z_pd_CA3_CA1=sig_z_pd[sig_z_pd['group']=='CA3 compared to CA1']['normalized U']
sig_z_pd_CA3_CA3=sig_z_pd[sig_z_pd['group']=='CA3 compared to CA3']['normalized U']
z,p=stats.ttest_ind(sig_z_pd_CA3_CA1,sig_z_pd_CA3_CA3)
print(f'the p value for this difference is {p}')
```

Fig 11. Distribution of Mann Whitney U values for all CA3 cells in each session where the distribution of r values for that cell is significantly different than expected. The normalized U value is obtained by dividing the U values from each test by the product of the size of the two population compared.   The normalized U value of the distrubution of r values between CA3 cells and local cells is significantly lower (indepent t test, p=7.515e-9) than the value for the distrubution of r values between CA3 and CA1 cells.  This indicates that the distribution of r values of CA3 cells with local cells is significantly more different than expected than the distribution of r values of CA3 cells with CA1 cells.

## Conclusion & Discussion 
  In this analysis we looked at the correlation of cells from CA3 to other regions in the hippocampus, and found some correlation between projection density and the level of activity correlation between regions. We showed that the CA3 region, which has higher local projection density compared to its projection density with CA1, also has significantly more correlation among local cells than correlation to CA1 cells.  Comparison of the distribution of correlation coefficients between neurons in each region, from the actual and randomly generated populations, allows us to visualize the extent of skewness in neuron firing patterns in CA1 and CA3. The more skew in the distribution, the higher the magnitude of the correlation coefficients. A correlation coefficient with a high magnitude indicates that the neurons in the regions being compared have related activity, be it excitatory or inhibitory. Most of the violin plots with significant p-values for the difference between correlation coefficient distributions have right skews, indicating that the firing pattern of neurons in one region varies directly with the firing pattern in the target region.
  
  Even though the projections inspected here are significantly better than those examined in our previous work with the cell types project, which had projection density around 1% as compared to 30% in this study, there are still flaws presented by the lack of related data across the Allen Connectivity and Neuropixel datasets.  Even though we were able to find plenty of recordings from cells from the CA3 and CA1 regions, recording from CA2 region is scarce, rendering the comparson to CA2 cells impossible.  Analysis of projection strength by fluorescence can also be problematic, even though our choice of examining the projection density does mitigate some uncertainty introduced by signal fluctuation across scans, by only analyzing the absense or presence of the fluorescent signal. One potential problem is that the level of fluorescence does not directly correlate with the number of synapses, as axons could pass by a region without making any synapses to its cells.  This kind of analysis is especially problematic with local projections, as the fluorescence of the orginal injection site must be manually identified and excluded.  Here we largly relied on identification provided by the Allen database, but such identification might be unreliable.  The local fluorescence level might also be naturally higher than fluorescence level in downstream areas, as it includes axons projecting to all downstream regions as wells as cell bodies and local dendrites.  This is a big problem for our analysis, especially since we didn't find enough CA2 cells, as our analysis largely hinged on comparing inter-region correlation verses intra-region correlation.  This increased correlation we found could also be driven by shared input of the local CA3 cells rather than being influenced by projection patterns.  Another potential downside of our study lies in using the Spearman's correlation test for examining activity correlation, as the influence of synaptic input might only increase correlation of activity by some time delay equal to the temporal dynamic of the synapses in question.  The selection of electrode placement found in the Allen dataset might also introduce bias to our analysis.  For example, the distal portions of CA3 connect more to CA1 than the proximal portions, and so if there is more data from distal portions, it would appear there is higher connectivity. Future studies should include a cross correlation method that includes an estimate of correlation across different time delays, and maybe use viral tracers that expresses synapse tagging proteins besides normal fluorescences.  Overall, this analysis examined the limit of the current set of data available from the Allen database for addressing certain type of problems, however, future data can be incorporated to make stronger comparisons available.
