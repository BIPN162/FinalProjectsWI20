```python
# get experiments where the viral injection landed in one of the area above
overlap = mcc.get_experiments(dataframe=True,injection_structure_ids = ids)

#for loop to limit search only to experiments where most of the virus is in the desired structures
# (not experiments where the structures were infected by accidental spread)
in_str_list = []
for id, exp in overlap.iterrows():
    primary_injection_structure=exp.primary_injection_structure
    #Append true if primary injection structure is in our list of structure ids
    in_str_list.append(primary_injection_structure in ids)
overlap = overlap[in_str_list]
```

```python
#creating digraph to store projections
str_name = overlap.structure_name.unique()  #get list of structures found in the overlap
stct_ids = overlap.structure_id.unique()    #get ids of structures found in the overlap
label={}                                    #create a dictionary of names for plotting algorithm
attr_lst=['id', 'acronym', '_name']         #list of attributes each node would have
Vctx_cnctn = nx.DiGraph()                   #create a graph with directed edge using networkx
for id in stct_ids:                         #iterating through each id found in overlap
    name = G.nodes[id]['_name']     #get structure fullname according to id
    Vctx_cnctn.add_node(name)       #add node to directed graph with name = structure fullname
    label[name] = name              #add to dictionary of names for plotting algorithm
    for attr in attr_lst:           #append id, acronym, name attribute to the created node
         Vctx_cnctn.nodes[name][attr] = G.nodes[id][attr]
```

```python
#code for modelling all recorded projections as DiGraph in networkx
print("%d total experiments" % len(overlap)) #print out number of total experiments in overlap
i = 0                                        #counter i set to 0
for id, exp in overlap.iterrows():           #iterate through each experiment in overlap
    id = [id]                                #id of the experiment, a list is required for the allensdk method
    cre = exp.transgenic_line                #cre line of this experiment
    spread = len(exp.injection_structures)   #number of structures that the virus spread to 
    if cre==None:
        cre = 'wildtype'
    str_id = exp.structure_id                #structure id of this experiment
    str_nm = G.nodes[str_id]['_name']        #get name of structure using ontology graph
    nd_nm = str_nm+'  '+cre                  #node name is set to the combination of structure name and cre name, each node //
                                             #will be a structure and cell type defined population
    exp_uni = mcc.get_structure_unionizes(experiment_ids = id)            #getting allen's characterization of projections from this experiment
    target = exp_uni[exp_uni['structure_id'].apply(lambda x: x in ids)]   #remove projections to structures that are not 
                                                                          #included in neuropixel data 
    target = exp_uni[exp_uni['projection_density'].apply(lambda x:x!=0)]   #remove projections to structures that 
                                                                           #have 0 projection density 
    target =  target[target['hemisphere_id']==3]                           #set to look at combined density of both hemisphere
    target_not_injection = target[~target['is_injection']]                 #focus on fluorescence that is labeled not injection site
    for _ , projection in target_not_injection.iterrows():                 #for each projection in focus
        prjctn_str = projection.projection_density /spread                 #projection strength(edge weight) is set to projection density recorded, divided by spread
        end_id = projection.structure_id                                   #end_id is id of structure where fluorescence is detected 
        if not end_id in ids:                                              #skip if a projection is not in list of structures
            continue
        end_str =  G.nodes[end_id]['_name']                                #name of end structure 
        Vctx_cnctn.add_node(nd_nm)                                         #create a new node for the structure and cell type defined population
        rooted = Vctx_cnctn.has_edge(str_nm,nd_nm)                         #the new node is rooted if there is an edge pointing from its parent structure to it
        connected = Vctx_cnctn.has_edge(nd_nm,end_str)                     #check if a projection from the population to the target structure already exist
        if not rooted:                                                     #create an edge point from parent structure to cell type defined population in that structure 
            Vctx_cnctn.add_edge(str_nm,nd_nm)
        if not connected:                                                  #create the new hypothetical projection if it does not yet exist
            Vctx_cnctn.add_edge(nd_nm,end_str,weight = prjctn_str)
            Vctx_cnctn[nd_nm][end_str]['sum'] = 1                          #create attribute sum
            Vctx_cnctn[nd_nm][end_str]['sum_w'] = prjctn_str               #create attribute sum of weight
            Vctx_cnctn[nd_nm][end_str]['weight'] = prjctn_str              #assign weight to weight attribute
        else:                                                              #if the projection already exist
            sum_weight = Vctx_cnctn[nd_nm][end_str]['sum_w']               
            sum_weight += prjctn_str
            Vctx_cnctn[nd_nm][end_str]['sum_w'] = sum_weight               #add new weight to sum weight
            Vctx_cnctn[nd_nm][end_str]['sum'] += 1                         #increase sum attribute by one
            pjt_sum = Vctx_cnctn[nd_nm][end_str]['sum']
            Vctx_cnctn[nd_nm][end_str]['weight'] = sum_weight/pjt_sum      # recalculate average weight from sum and sum weight
    i+=1                                                                   #increment counter
    if i%5 == 0:                                                           #report progress every for every 5 exp analized 
        print(f'{i} experiments analyzed')
```

```python
#code to put nodes on concentric circles, parent regions will be on the outer circle while region+celltype defined population will be on the inner cicle
def get_pos(G):   
    pos = nx.circular_layout(G)                    # code to align all nodes on a circle
    angs = np.linspace(0, 2*np.pi, 1+len(ids))     #len(ids) equally spaced phase from 0 to 2pi 
    repos = []                                     #list for new position for each node
    rad =  6                                       # radius of circle
    for ea in angs:
            repos.append(np.array([rad*np.cos(ea), rad*np.sin(ea)]))  # generate position of len(ids) points on the new circle
    i = 0                                       #counter
    for strct in fullname:                      #put parent structures on outer circle
        pos[strct] = repos[i]
        i+=1
        
    angs = np.linspace(0, 2*np.pi, 1+len(ids))  #same procedure but for inner circle
    repos = []
    rad = 3                                     # radius of  inner circle
    for ea in angs:
            repos.append(np.array([rad*np.cos(ea), rad*np.sin(ea)]))
    for ea in pos.keys():                       # putting nodes with cre in the name on the inner circle near it's parent with some random jitter
        isCre = 'Cre' in ea
        if not isCre:
            continue
        i = 0
        for strct in fullname:
            if strct in ea:
                pos[ea] = repos[i]+(1.5*np.random.rand(2)-0.75)
                break
            i+=1
    return pos
```

```python
#function for turning spike time data to fire rate
def get_st(lst):
    #lst, list of unit ids
    st = []
    #append spike time of units with the id in lst to st
    for id in lst:
        st.append(session.spike_times[id])
    return st
```

```python
#function for turning spike time data to fire rate
def bin_st(bin_size,start,end,st):
    #bin_sizeL size of time bin to compute firing rate
    #start: start time of trial
    #end: end time of trial
    #st: spike time
    #duration of trial
    duration = end - start
    #number of units in spike time list
    numOfUnits = len(st)
    #number of bins for this trial
    numOfBins= int(round(duration/bin_size))
    #x is the container for binned firerate
    x = np.zeros((numOfUnits,numOfBins))
    #increase spike count of a time bin by one if a spike time land in a perticular time bin
    for i in range(len(st)):
        tmp = st[i][(st[i]>=start) & (st[i]<=end)]
        for j in range(len(tmp)):
            bin_id = int(round((tmp[j]-start) /bin_size)-1)
            x[i][bin_id]+=1
    return x
```

```python
#function to find correlation distribution between rows of two matrixes with uneven number of rows, np.corrcoef can only deal with arrays with same shape
def find_corr(x,y):
    #chect if x and y are the same matrix
    same_area = np.array_equal(x,y)
    #containrer for row correlation matrix
    ans = np.zeros((len(x),len(y)))
    #for each row in x
    for i in range(len(x)):
        #set ith row in x to xi
        xi = x[i]
        y_=y
        #skip correlation between same rows if x and y are the same matrix
        if same_area:
            y_=np.delete(y,i,0)
        #for each row in y except the same row if x and y are the same
        for j in range(len(y_)):
            #set jth row in y to yj
            yj = y_[j]
            #find correlation bewteen ith row and jth column in x and y, store to the output matrix
            ans[i][j] = np.corrcoef(xi,yj)[0][1]
    return ans        
```

```python
#function to round to sigfig
from math import log10, floor
def round_sig(x, sig=2):
    return round(x, sig-int(floor(log10(abs(x))))-1)
```

```python
#taking a look at session 744228101, the first session in all relevant sessions
ids = 744228101

nwb_path = f'C:\\allen data\\ecephys_session_{ids}.nwb'
session = EcephysSession.from_nwb_path(nwb_path, api_kwargs={
        "amplitude_cutoff_maximum": np.inf,
        "presence_ratio_minimum": -np.inf,
        "isi_violations_maximum": np.inf
    })
#get all units from this session with signal to noise ratio bigger than 2
units = session.units[session.units['snr'] > 2]
```

```python
#find epoch of natural movies
#get all stimulus epochs of the session
epochs = session.get_stimulus_epochs()
#limit to only stimulus type: natural_movie_three, which is a 600 second movie
epochs = epochs[epochs.stimulus_name =='natural_movie_three']
#assign start and end time of all trials of natural_movie_three
start = list(epochs.start_time)
end = list(epochs.stop_time)
```

```python
#get id of units in CA1,CA2,CA3 respectively
CA1_ids = units[units['ecephys_structure_acronym']=='CA1'].index.values
CA2_ids =units[units['ecephys_structure_acronym']=='CA2'].index.values
CA3_ids =units[units['ecephys_structure_acronym']=='CA3'].index.values
print(f'There are {len(CA1_ids)} CA1 units in this session')
print(f'There are {len(CA2_ids)} CA2 units in this session')
print(f'There are {len(CA3_ids)} CA3 units in this session')
```

```python
#CA2 only has one unit, we will skip it in the following analysis
CA1_st = get_st(CA1_ids)
CA3_st = get_st(CA3_ids)
```

```python
#bin the spike time data and average firing rate across trials
bin_size = 1
#number of trials with natural_movie_three
num_of_tr=len(start)
#prepare three np array with zeros and the correct size (rows = number of neurons in region, columns = length of movie / bin size)
CA1_fr=np.zeros((len(CA1_ids),round(600.50163/bin_size)))
CA3_fr = np.zeros((len(CA3_ids),round(600.50163/bin_size)))
#bin the spike time data for CA1,2 and 3 and add up all trials in np array
for i in range(num_of_tr):
    strt = start[i]
    nd = end[i]
    CA1_fr+=bin_st(bin_size,strt,nd,CA1_st)
    CA2_fr+=bin_st(bin_size,strt,nd,CA2_st)
    CA3_fr+=bin_st(bin_size,strt,nd,CA3_st)
#average of firing rate across trial and remove all cells that did not fire
CA1_fr = CA1_fr/num_of_tr
CA3_fr = CA3_fr/num_of_tr
CA1_fr= CA1_fr[~np.all(CA1_fr == 0,axis=1)]
CA3_fr= CA3_fr[~np.all(CA3_fr == 0,axis=1)]
#average of firing rate across neurons
CA1_avg= np.average(CA1_fr,axis = 0)
CA3_avg= np.average(CA2_fr,axis = 0)
```

```python
#find correlations between units in CA3 and CA1 or within CA3
corr_CA3_CA1=pd.DataFrame(np.transpose(find_corr(CA3_fr,CA1_fr))).dropna()
corr_CA3_CA3=pd.DataFrame(np.transpose(find_corr(CA3_fr,CA3_fr))).dropna()
```

```python
#average firing rate for each bin for CA1 and CA3 cells
CA1_bin_avg = np.average(CA1_avg)
CA3_bin_avg = np.average(CA3_avg)
```

```python
#generate random firing rate for independent 'neurons' 
num_of_bin=len(CA1_avg)
rand_CA3=np.zeros((1,num_of_bin))
rand_CA3[0] = np.random.poisson(lam = CA3_bin_avg,size = num_of_bin)

def rand_n_neurons(average,n_neurons,num_of_bin):
    rand_matrix = np.zeros((n_neurons,num_of_bin))
    for i in range(n_neurons):
        rand_matrix[i]=np.random.poisson(lam = average,size = num_of_bin)
    return rand_matrix

rand_CA1_100 = rand_n_neurons(CA1_bin_avg,100,num_of_bin)
rand_CA3_100 = rand_n_neurons(CA3_bin_avg,100,num_of_bin)
rand_CA1_500 = rand_n_neurons(CA1_bin_avg,500,num_of_bin)
rand_CA3_500 = rand_n_neurons(CA3_bin_avg,500,num_of_bin)
rand_CA1_1000 = rand_n_neurons(CA1_bin_avg,1000,num_of_bin)
rand_CA3_1000 = rand_n_neurons(CA3_bin_avg,1000,num_of_bin)

f,ax=plt.subplots(1,3)

#find firing correlations between a randomly generated 'CA3' with 100 random CA1 or CA3 units
rand_corr_CA3_CA1=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA1_100))).dropna()
rand_corr_CA3_CA3=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA3_100))).dropna()
sns.violinplot(data=rand_corr_CA3_CA1, palette="Set3", bw=.5, cut=1, linewidth=1,ax=ax[0])

rand_corr_CA3_CA1=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA1_500))).dropna()
rand_corr_CA3_CA3=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA3_500))).dropna()
sns.violinplot(data=rand_corr_CA3_CA1, palette="Set3", bw=.5, cut=1, linewidth=1,ax=ax[1])

rand_corr_CA3_CA1=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA1_1000))).dropna()
rand_corr_CA3_CA3=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA3_1000))).dropna()
sns.violinplot(data=rand_corr_CA3_CA1, palette="Set3", bw=.5, cut=1, linewidth=1,ax=ax[2])
```

```python
rand_CA3_n=rand_n_neurons(CA1_bin_avg,len(CA3_fr),num_of_bin)
rand_CA1_n=rand_n_neurons(CA1_bin_avg,len(CA3_fr),num_of_bin)
rand_corr_CA3_CA1=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA1_n))).dropna()
rand_corr_CA3_CA3=pd.DataFrame(np.transpose(find_corr(rand_CA3,rand_CA3_n))).dropna()
sns.violinplot(data=rand_corr_CA3_CA1, palette="Set3", bw=.5, cut=1, linewidth=1)
```

```python
#plot distrubution of correlation coefficients of all CA3 cells with other cells in CA1
f, ax = plt.subplots(figsize=(25, 6))

# Draw a violinplot for histogram of each CA3 cells 
sns.violinplot(data=corr_CA3_CA1, palette="Set3", bw=.5, cut=1, linewidth=1)

#display p value for CA3 cells whose distribution is significantly skewed
# Calculate p value and z score for skewness test of the distribution of coefficients of each CA3 neuron with neurons in CA1
p_value =np.apply_along_axis( stats.skewtest, axis=1, arr=corr_CA3_CA1 )[:,1]
z_score =np.apply_along_axis( stats.skewtest, axis=1, arr=corr_CA3_CA1 )[:,0]

#create a list of string to show if p value is significant
p_value_text = [str(round_sig(x,3)) for x in p_value.tolist()]
p_value_text = ["*p: " + i for i in p_value_text]

#find position of each label
y_pos = np.ones(len(corr_CA3_CA1))*-1
x_pos = range(len(p_value_text))
for tick,label in zip(pos,ax.get_xticklabels()):
    if p_value[tick]<0.05:
        ax.text(x_pos[tick], y_pos[tick], p_value_text[tick], horizontalalignment='center', size='large', weight='semibold',rotation=30)

# Finalize the figure
ax.set(ylim=(-1, 1))
sns.despine(left=True, bottom=True)
ax.set_title('correlation coefficient between CA3 cells and CA1 cells',fontsize='x-large')
ax.set_ylabel('correlation coefficient',fontsize='x-large')
ax.set_xlabel('Cells in CA3',fontsize='x-large')
ax.tick_params(axis='both', which='major', labelsize=20)
```
