# Background and Prior Work

The hippocampus is one of the most studied neuronal systems in the brain. Located in the medial temporal lobes, the hippocampus is responsible for forming long term memory (Bird and Burgess, 2008). Inside the hippocampus there are four subregions forming the neural circuit, including CA1 region and CA3 region. Both regions contain densely packed pyramidal neurons, which integrate input information and outputs to other brain areas (Ginsberg and Che 2005). Interestingly, the two regions have distinct patterns of connectivity, which match with their differential functions in retrieval of episodic contexts (Dimsdale-Zucker et al. 2018). We are motivated to study both the difference between the general picture -- CA1 and CA3 neurons, and also a specific neuron type in those regions -- pyramidal neurons in CA1 and CA3 to see if those neurons’ electrophysiological properties match with the functional variation of the two areas, and if gene expression can explain such difference in electrophysiology. The electrophysiological features are interesting to look at because they describe the essential activity of a neuron, such as the spiking patterns, frequency of action potentials, etc.

To date, many papers have worked to characterize the CA1 and CA3 cell types in order to understand how their properties produce the functional circuit. Dating back to the 1990’s, work was being done to understand the unique electrophysiological and morphological properties of CA3 neurons (Bilkey and Schwartzkroin 1990). With the advancements in high throughput genomics, it was revealed that the CA3 and CA1 regions have many differences in gene expression of glutamate and GABA receptors, cytoskeletal elements, and even immediate early genes which play an important role in regulating synaptic strength (Ginsberg and Che 2005). Extensive work has also been to characterize the variations of electrophysical properties within the subregions of both CA3 and CA1 (Milior et al. 2016). However, not much work has been done correlating the unique gene expression of each region with their electrophysiological properties. High throughput methods producing large, public, datasets has made it relatively easy in recent years to investigate such relationships. One paper, looked at the correlation between gene expression and electrophysiology features across a wide range of cell types and brain regions (Tripathy et al. 2017). They identify over 400 genes whose expression correlate well with an electrophysiology feature. However they did not look in depth at the difference in gene expression between more closely related regions like the CA3 and CA1. Here, we investigate if the same methodology can be used to correlate differences in gene expression with differences in electrophysiological properties. To do this we will use a similar method to Tripathy et al. by analyzing gene expression data from Allen Brain Atlas RNA-Seq dataset and electrophysiological from NeuroElectro Project dataset. 

The NeuroElectro Project is a centralized database that extracts published cell electrophysiological properties recordings from literature. There are a total of 1515 observations, 352 columns including general information about the paper, species, tested neuron type, and the electrophysiological properties such as resting membrane potential, input resistance, etc. One thing to note is this dataset has many NA observations, as in the paper published not all electrophysiological features are taken care of at a time. After filtering for rodents hippocampal CA1 and CA3 pyramidal cells, we found 111 observations for CA1, and 25 observations for CA3. In total we have 154 observations for all CA1 neurons, and 39 observations for all CA3 neurons. These observations were later used to compare the electrophysiological properties between two neuron types.

The second dataset was the Allen Brain Atlas Cell Types Database: RNA-Seq Data. This database is specifically for cell diversity in the mouse cortex and hippocampus, which consists of RNA-Seq results from close to 75,000 single cells from multiple cortical areas and the hippocampus. The amount of RNA represents the gene enrichment in the cell. In total, there are 45768 genes (rows) and 290 clusters of neurons in various mouse brain regions (columns). The dataset is used in our analysis for filtering out RNA that are expressed in CA1 and CA3, in which we further isolate ones that correspond to specific electrophysiological features.

## Hypothesis

Based on the different functionalities of CA1 and CA3 regions and specifically the pyramidal cells, the electrophysiological features of the two neuron types will most likely differ: for example, the voltage difference between resting and active state, the time to return to resting state after firing action potential, etc. Also, a lot of electrophysiological activities of neurons are achieved through opening and closure of voltage-gated ion channels, or regulatory factors binding to those channels. Since the enrichment of RNA in cells is a good indicator for abundance of corresponding proteins, we hypothesize that for neurons that have significantly higher frequency of electrophysiological activities, a higher level of gene expression for genes that controls the open/close of ion channels would be observed.

## Data Analysis<a name="DataAnalysis"></a>

We aimed to use the p-value obtained from two sample Student t-test to determine whether the features/gene expression were different between CA1 and CA3 regions. The alpha level was set at 0.05. However, two sample t-test has an assumption that the samples are drawn from a normal distribution. Therefore, we decided to perform the skewness test to verify the usage of t-tests. 

We used the [stats.skewtest() method](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.skewtest.html#scipy.stats.skewtest), which implements the D'Agostino-Pearson skewness test to test for normality of the distribution. The null hypothesis is that the sample is drawn from a normal distribution, so a smaller p-value indicates a higher likelihood that the distribution is skewed. We used it for all neurons in CA1 and CA3 regions, and specifically pyramidal cells in CA1 and CA3. 


***Note***: we have a relatively small dataset for electrophysiological features: for pyramidal cells, we have 35 observations for CA1 and 10 for CA3; and even if we combine all, the dataset is still small. Because of this, we made two adjustments in our following analysis:
 
- Since skew test only works for sample size to be at least 8, we decided to exclude features that have fewer than 8 observations in following analysis, as a small sample size could be largely affected by random noise.
 
- Because of the nature of our dataset, we decided to only exclude really skewed points, which are the features that have p-value < 0.005. Using this adjustment, we decided to tolerate a higher level of skewness to draw conclusions from this dataset. 


After using the screening result of the skewness test, for electrophysiological features that are both normal (p>0.005) for CA1 and CA3 neurons, we performed the two sample Student t-test. For features that are skewed in either CA1 or CA3, or both of them, we performed the Mann-Whitney U test for unpaired samples.

```python
# Create a list to only store the column names that store data to analyze. This is to faciliate the analysis on the
# next step. 
data_col = []

col = list(CA1_ephys_df.columns)

for i in col:
    if 'sd' in i:
        continue
    elif '_raw' in i:
        continue
    elif '_err' in i:
        continue
    elif "_n" in i:
        continue
    else:
        data_col.append(i)
```

### For CA1 and CA3 cells overall: <a name="t-test"></a>
We first checked the skewness of electrophysiological features for all CA1 and CA3 neurons. 

```python
skewness_CA3_skewtest_all = []
skewness_CA1_skewtest_all = []

# For each electrophysiological feature, if the amount of available data points is less than 8, 
# continue to next feature. If not, perform the skewness test. If the returned p-value is greater 
# than 0.005, the data points are normal. If not, the data points are skewed. 
# Note: Skewness test only works for more than 8 data points 
for a in data_col:
    sample_size = len(CA3_all_df) - CA3_all_df[a].isnull().sum()
    if (sample_size < 8):
        skewness_CA3_skewtest_all.append(a + ': sample size is ' + str(sample_size) + ', too small to perform skewness test')
        continue
    else:
        CA3_skewed_stats_all, CA3_skewed_all = stats.skewtest(CA3_all_df[a],nan_policy='omit')
        if CA3_skewed_all < 0.005:
            skewness_CA3_skewtest_all.append('skewed')
        elif CA3_skewed_all > 0.005:
            skewness_CA3_skewtest_all.append('normal')

# Same operation for CA1 electrophysiological features. 
for b in data_col:
    sample_size = len(CA1_all_df) - CA1_all_df[b].isnull().sum()
    if (sample_size < 8):
        skewness_CA1_skewtest_all.append(b + ': sample size is ' + str(sample_size) + ', too small to perform skewness test')
        continue
    else:
        CA1_skewed_stats_all, CA1_skewed_all = stats.skewtest(CA1_all_df[b],nan_policy='omit')
        if CA1_skewed_all < 0.005:
            skewness_CA1_skewtest_all.append('skewed')
        elif CA1_skewed_all > 0.005:
            skewness_CA1_skewtest_all.append('normal')
            
# Sample Output
skewness_CA1_skewtest_all
```

The skewness test shows that about half of electrophysiological features inside this dataset is too small; besides, it indicates that we need to take care of some skewed features in following analysis. We then performed t-test and U test accordingly. 

```python
# Initialize lists to store each value 
test_results_ephys_all = []
p_value = 0
p_value_before_transformation_all = []
feature_all = []

# For each electrophysiological feature, if the feature datasets on both regions (CA1 and CA3) were determined 
# as ‘normal’ from the previous test, we will perform t-test to determine significance. If one of them was 
# determined as 'skewed', perform Mann Whitney test for them. 
for i in range(len(data_col)):
    if (skewness_CA3_skewtest_all[i] == 'normal') and (skewness_CA1_skewtest_all[i] == 'normal'):
        p_value = stats.ttest_ind(CA3_all_df[data_col[i]],CA1_all_df[data_col[i]],nan_policy='omit').pvalue
        test_results_ephys_all.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by 2 sample t-test')
        p_value_before_transformation_all.append(p_value)
        feature_all.append(data_col[i])
        # print(data_col[i],p_value) 
        
    elif skewness_CA3_skewtest_all[i] == 'skewed':
        if skewness_CA1_skewtest_all[i] == 'skewed':
            p_value = stats.mannwhitneyu(CA3_all_df[data_col[i]],CA1_all_df[data_col[i]]).pvalue
            test_results_ephys_all.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation_all.append(p_value)
            feature_all.append(data_col[i])
            # print(data_col[i],p_value) 
        elif skewness_CA1_skewtest_all[i] == 'normal':
            p_value = stats.mannwhitneyu(CA3_all_df[data_col[i]],CA1_all_df[data_col[i]]).pvalue
            test_results_ephys_all.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation_all.append(p_value)
            feature_all.append(data_col[i])
            # print(data_col[i],p_value)
        else:
            continue
    
    elif skewness_CA1_skewtest_all[i] == 'skewed':
        if skewness_CA3_skewtest_all[i] == 'skewed':
            p_value = stats.mannwhitneyu(CA3_all_df[data_col[i]],CA1_all_df[data_col[i]]).pvalue
            test_results_ephys_all.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation_all.append(p_value)
            feature_all.append(data_col[i])
            # print(data_col[i],p_value)
        elif skewness_CA3_skewtest_all[i] == 'normal':
            p_value = stats.mannwhitneyu(CA3_all_df[data_col[i]],CA1_all_df[data_col[i]]).pvalue
            test_results_ephys_all.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation_all.append(p_value)
            feature_all.append(data_col[i])
            # print(data_col[i],p_value)
        else:
            continue
            
test_results_ephys_all
```

The result of the two-sample t-test and Mann-Whitney U test in respect to electrophysiological features is illustrated above. None of the raw p-values pass the threshold alpha = 0.05. 

In addition, in order to counteract the problem of multiple comparisons, we apply the [Bonferroni Correction](https://en.wikipedia.org/wiki/Bonferroni_correction), and comparing each p-value to the corrected threshold alpha = 0.05/ # of trials $\approx$ 0.006.

```python
# Perform the Bonferroni Correction to adjust the alpha level. Print out electrophysiological feature
# that pass the new alpha level. 
alpha = 0.05
trials = len(p_value_before_transformation_all)
alpha_corrected = alpha / trials 

for i in range(len(feature_all)):
    if p_value_before_transformation_all[i] < alpha_corrected:
        print(str(feature_all[i]) + " is significantly different between CA1 and CA3 after Bonferroni Correction.")
        
# No features are printed after the correction, showing all of the previously selected features are not significant.
```

**Interpretation**: we found that none of our electrophysiological features pass the threshold, meaning none of the electrophysiological features are significantly different across all CA1 and CA3 neurons in rodents. This can be explained by the fact that there are different neurons inside those two regions, some are excitatory, and some are inhibitory. When we combine all neurons, each electrophysiological feature in each region will end up having higher variance due to the difference between neuron types, so that it is harder to pass statistical tests. We decided to look at a pair of more specific cell type in each region in order to find the differences. 

### For CA1 and CA3 pyramidal cells only:

We then took a deeper look at the pyramidal cells in CA1 and CA3 regions, which is also the initial analysis we did in the cell type project. We wanted to confirm if the same result would hold true for all rodents, not just mice. We first checked the skewness of electrophysiology features for CA1 and CA3 pyramidal cells, and did two sample student t test and Mann-Whitney U test accordingly.

```python
skewness_CA3_skewtest = []
skewness_CA1_skewtest = []

# Skewness test only works for more than 8 data points 
        
for a in data_col:
    sample_size = len(CA3_df) - CA3_df[a].isnull().sum()
    if (sample_size < 8):
        skewness_CA3_skewtest.append(a + ': sample size is ' + str(sample_size) + ', too small to perform skewness test')
        continue
    else:
        CA3_skewed_stats, CA3_skewed = stats.skewtest(CA3_df[a],nan_policy='omit')
        if CA3_skewed < 0.005:
            skewness_CA3_skewtest.append('skewed')
        elif CA3_skewed > 0.005:
            skewness_CA3_skewtest.append('normal')
            
for b in data_col:
    sample_size = len(CA1_df) - CA1_df[b].isnull().sum()
    if (sample_size < 8):
        skewness_CA1_skewtest.append(b + ': sample size is ' + str(sample_size) + ', too small to perform skewness test')
        continue
    else:
        CA1_skewed_stats, CA1_skewed = stats.skewtest(CA1_df[b],nan_policy='omit')
        if CA1_skewed < 0.005:
            skewness_CA1_skewtest.append('skewed')
        elif CA1_skewed > 0.005:
            skewness_CA1_skewtest.append('normal')
skewness_CA1_skewtest
```

```python
test_results_ephys = []
p_value = 0
p_value_before_transformation = []
feature = []

# For each electrophysiological feature, if the feature datasets on both regions (CA1 and CA3) were determined 
# as ‘normal’ from the previous test, we will perform t-test to determine significance. If one of them was 
# determined as 'skewed', perform Mann Whitney test for them. 
for i in range(len(data_col)):
    if (skewness_CA3_skewtest[i] == 'normal') and (skewness_CA1_skewtest[i] == 'normal'):
        p_value = stats.ttest_ind(CA3_df[data_col[i]],CA1_df[data_col[i]],nan_policy='omit').pvalue
        test_results_ephys.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by 2 sample t-test')
        p_value_before_transformation.append(p_value)
        feature.append(data_col[i])
        # print(data_col[i],p_value) 
        
    elif skewness_CA3_skewtest[i] == 'skewed':
        if skewness_CA1_skewtest[i] == 'skewed':
            p_value = stats.mannwhitneyu(CA3_df[data_col[i]],CA1_df[data_col[i]]).pvalue
            test_results_ephys.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation.append(p_value)
            feature.append(data_col[i])
            # print(data_col[i],p_value) 
        elif skewness_CA1_skewtest[i] == 'normal':
            p_value = stats.mannwhitneyu(CA3_df[data_col[i]],CA1_df[data_col[i]]).pvalue
            test_results_ephys.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation.append(p_value)
            feature.append(data_col[i])
            # print(data_col[i],p_value) 
        else:
            continue
    
    elif skewness_CA1_skewtest[i] == 'skewed':
        if skewness_CA3_skewtest[i] == 'skewed':
            p_value = stats.mannwhitneyu(CA3_df[data_col[i]],CA1_df[data_col[i]]).pvalue
            test_results_ephys.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation.append(p_value)
            feature.append(data_col[i])
            # print(data_col[i],p_value) 
        elif skewness_CA3_skewtest[i] == 'normal':
            p_value = stats.mannwhitneyu(CA3_df[data_col[i]],CA1_df[data_col[i]]).pvalue
            test_results_ephys.append(str(data_col[i]) + ' raw pvalue = ' + str(p_value) + ' by Mann Whitney U Test')
            p_value_before_transformation.append(p_value)
            feature.append(data_col[i])
            # print(data_col[i],p_value)
        else:
            continue
test_results_ephys  
```

The result of the two-sample t-test and Mann-Whitney U test in respect to electrophysiological features is illustrated above. Only the raw p-value of tau, the time constant, passes the threshold alpha = 0.05. To counteract the problem of multiple comparisons, we apply the Bonferroni Correction, and comparing each p-value to the corrected threshold alpha = 0.05/ # of trials $\approx$ 0.0033.

```python
# Perform Boneffroni Correction to adjust the alpha level in order to prevent type I error
alpha = 0.05
trials = len(p_value_before_transformation)
alpha_corrected = alpha / trials 

# Compare the p-value result with the new corrected alpha level, print out the features that could pass the test
for i in range(len(feature)):
    if p_value_before_transformation[i] < alpha_corrected:
        print(str(feature[i]) + " is significantly different between CA1 and CA3 after Bonferroni Correction.") 
```

**Interpretation**: Since only the p-value of tau is smaller than the significance level = 0.003 after Bonferroni correction, it is the only feature that is significantly different between the two neurons, while others are either not possible to compare or not significantly different. It is even able to pass a stricter threshold of alpha = 0.005, which is consistent with the significance level we picked for skewness test. We then picked tau, the time constant, to analyze the genetic contribution to the difference. 

From the paper (Tripathy, Shreejoy et al. 2017), we obtained four genes that were demonstrated to contribute to the tau property of neurons. They are *Kcnab2*, *Itpr1*, *Kcna1*, and *HCn3*. 
 
This time we used a different skewness test, the [Shapiro-Wilk test](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html) to test for normality. We decided to use a different skewness test here because most of the gene expression data have less than 8 data points. The Shapiro-Wilk test is capable to work with data that has less than 8 data points and therefore is appropriate for the gene expression data. The null hypothesis is that the data was drawn from a normal distribution. So if the p-value is less than the alpha level, it means that the dataset is skewed and vice versa. If the gene expression for CA1 and CA3 is both normal, we performed two sample t-test, and if one of them is skewed, we used Mann-Whitney test. After that, we used Bonferroni correction to determine whether the gene expression level of these genes are indeed different in both neurons.


Reference:
Tripathy, Shreejoy J et al. [“Transcriptomic correlates of neuron electrophysiological diversity.”](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5673240/) PLoS computational biology vol. 13,10 e1005814. 25 Oct. 2017, doi:10.1371/journal.pcbi.1005814

```python
gene_list = [CA3_Kcnab2,CA1_Kcnab2,CA3_Itpr1,CA1_Itpr1,CA3_Kcna1,CA1_Kcna1,CA3_Hcn3,CA1_Hcn3]
skewed_p_value = []

# If the sample size has less than 3 data points, print out 'too small dataset'. If not, perform the shapiro-wilk test.
for item in gene_list:
    sample_size = len(item) - item.isnull().sum()
    if sample_size < 3:
        skewed_p_value.append('too small dataset')
        continue
    else:
        skewed_stats, skewed_p = stats.shapiro(item)
        skewed_p_value.append(skewed_p)

print(skewed_p_value)        

```

```python
p_Kcnab2 = stats.ttest_ind(CA3_Kcnab2,CA1_Kcnab2).pvalue
p_Itpr1 = stats.mannwhitneyu(CA3_Itpr1,CA1_Itpr1).pvalue
p_Kcna1 = stats.ttest_ind(CA3_Kcna1,CA1_Kcna1).pvalue

print('The P-value for Kcnab2 between two neuron types is ' + str(p_Kcnab2))
print('The P-value for Itpr1 between two neurons types is ' + str(p_Itpr1))
print('The P-value for Kcna1 between two neurons types is ' + str(p_Kcna1))
print('not enough info for Hcn3')
print('###############################################')

# Bonferroni correction:
alpha = 0.05
trials = 3
alpha_corrected = alpha/trials

p_genes = [p_Kcnab2,p_Itpr1,p_Kcna1]
list_of_genes =['Kcnab2','Itpr1','Kcna1']

for i in range(len(p_genes)):
    if p_genes[i] < alpha_corrected:
        print(str(list_of_genes[i]) + " is significantly different between CA1 and CA3 neurons after bonferroni correction")
    
```

**Interpretation**:

The t-test and U test result for the gene expression suggests that *Itpr1* and *Kcna1* have significantly different expression levels across the two neuron types: *Itpr1* is significantly higher in CA1, while *Kcna1* is significantly higher in CA3. This result is consistent with the plots generated in the data visualization section, where CA1 has a higher *Itpr1* gene expression than CA3, and CA3 has a higher *Kcna1* gene expression than CA1. 

Combining the statistical analysis we’ve done above, we observe that the gene expression difference matches up with the disparity of electrophysiological features. Since gene expression level could potentially account for the feature difference, we will discuss the actual functions of those genes in the conclusion section.

## PCA and K-means Clustering for Selected Electrophyiological and Genetic Features <a name="PCA"></a>

Through the statistical analysis above, we determined that there are significant differences on the tau feature of both neurons and the gene expression of *Itpr1* and *Kcna1* have significant differences as well. We then want to see if the differences are significant enough that PCA could successfully produce two clusters based on the neuron type (CA1 pyramidal cell and CA3 pyramidal cell). 

Principal Component Analysis is a statistical method to reduce a high-dimensional feature matrix into a lower dimension in order to simplify the feature complexities. It is done through computing the covariance matrix between two variables. The eigenvectors indicate the principal components, while the eigenvalues quantifies the percent of variance (also known as the factor values). In other words, the higher the eigenvalues is, the more variance exists between two variables. (Joliffe, 2002) At the end, the principle components would just be orthogonal to each other. Since there are no similarities between each principle components, these components will be used to separate distinct groups in the samples. Note that there will be no labeling provided to the PCA model, so it is essentially an unsupervised learning -- it uses the clues from analyzed features to group the samples without any training. 

K-means clustering, on the other hand, separates the sample by a given amount of cluster while attempting to achieve optimal grouping, meaning that each cluster has their own center. The model repeatedly groups the samples in various ways that the error, or the linear distance between each data point and the cluster center, is minimized.(Forgy, 1965) Since the expected amount of cluster is pre-given to the model, it does involve some degrees of subjectivity but mostly it is still an unsupervised machine learning model.

In our project, we decided to use both models to first verify the result of grouping and second, to confirm whether the differences are significant enough that an unsupervised machine learning model could accurately distinguish both neuron types. 



***Reference***

(1) Jolliffe I.T. Principal Component Analysis, Series: Springer Series in Statistics, 2nd ed., Springer, NY, 2002, XXIX, 487 p. 28 illus. ISBN 978-0-387-95442-4

(2)  Forgy, Edward W. (1965). "Cluster analysis of multivariate data: efficiency versus interpretability of classifications". Biometrics. 21 (3): 768–769. JSTOR 2528559

### Performing PCA

```python
# Import package for PCA, K-means clustering method, and seaborn plotting packages
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import seaborn as sns

# Combine CA1 and CA3 neuron dataframe and drop the label column
all_sample_set = Combined_CA1_df.append(Combined_CA3_df)
all_sample_feature = all_sample_set.drop('NeuronName', axis = 1)
all_sample_feature
```

```python
# Initialize a pca and perform pca on the normalized dataset
pca = PCA()  
training_norm_feature = (all_sample_feature - all_sample_feature.mean())/all_sample_feature.std()
training_2D = pca.fit_transform(training_norm_feature)
```

```python
# Store PC1 and PC2 data and plot the clustering
all_sample_set['PC1'] = training_2D[:, 0]
all_sample_set['PC2'] = training_2D[:, 1]
sns.lmplot("PC1", "PC2", hue = 'NeuronName', data=all_sample_set, fit_reg=False)
```

The plot above illustrates that the pca has successfully separate the neuron types with first two principal componenets as new features. The number of neuron type in each group matches with the numebr of each neuron type in the label column as well. The result looks promising! 

We are interested to see which feature contributes the most to separate the neuron types. Hence we plotted a heatmap to visualize their influences on PC1 and PC2. 

```python
plt.matshow(pca.components_[0:2, :],cmap='viridis')
plt.yticks([0,1],['PC1','PC2'],fontsize=10)
plt.colorbar()
plt.xticks(range(len(all_sample_feature.columns)),all_sample_feature.columns,rotation=65,ha='left')
plt.tight_layout()
plt.show()
```

The plot illustrates the level of influences (i.e. the factor value) of each feature on the PCs. Since the level of influences are measured in absolute values, the brightest and the darkest boxes indicate the most influential features to the PCs. As shown in the plot, *Kcnb2, Kcna1,* and *Scn8a* influences the most to first and second principle components. In other words, these features help the most in separating the two neuron types. It is interesting to see that the gene expression levels here help to separate the neuron type more than the electrophysiological feature. This could be explained by the fact that gene expression level describes more detail differences than an electrophysiological feature. 

To further demonstrate if the PCA model accurately separates the neuron types, we perform a k-means clustering to confirm our result. 

```python
# initialize a k-means model with 2 components. Fit the feature set and plot 
# the clustering. 
kmeans = KMeans(n_clusters = 2)
kmeans.fit(training_2D)
all_sample_set['kmeans_2'] = kmeans.predict(training_2D) 
sns.lmplot("PC1", "PC2", hue = 'kmeans_2', data=all_sample_set, fit_reg=False)
```

The k-means clustering produced an identical plot as above. This proves that they both accurately separate the samples into two neuron types. 

# III. Conclusion & Discussion <a name="Conclusion"></a>

Overall, our results are consistent with our initial analysis that there are significant electrophysiological variations between CA1 and CA3 pyramidal cells in rodents, which is supported by differences in gene expression level. On the other hand, by our PCA and k-means clustering models, we have good reasons to believe that those two features could provide some predictive powers to tell the difference between two neuron types. 

For the electrophysiological features, we applied skewness test and Bonferroni correction in order to improve our statistical analysis of CA1 and CA3 pyramidal cells from the previous cell type project. Aside from the comparison between two neuron types (pyramidal cells in CA1/CA3 region), we also attempted to extrapolate comparison results of all neurons in those two regions. However, the two regions were proved to have no significant differences. One possible reason would be that the pattern differences might not be as obvious as for pyramidal cells only, since they are composed of many cells in the region. As we took an average view of the electrophysiological features within each region, the difference in different neuron groups contribute to a higher variance across the region, which makes the test hard to pass. After all, since the extrapolation for all neurons within the region is not successful for rodents, we narrowed down our target to pyramidal cells to see if we can find some differences. For pyramidal cells, initially we thought there would be a lot more significant ones due to the distinct connectivity and functionality of two neuron types. However, it turned out that only membrane time constants (tau) are statistically significant after correction. Tau describes the duration that a cell takes from action potential to resting state. The variation of this feature determines the neuron’s firing rate and how long an action potential lasts for a neuron, which may relate to the functionality of the neuron. 

For gene expression difference, initially we proposed genes code for voltage-gated ion channels that would contribute to the electrophysiological differences, and our result turns out to match with our prediction. After examination genes that are related to tau (Triphary et. al, 2017), we discovered Itpr1 and Kcna1 are significantly different among CA1 and CA3, where Itpr1 has a higher expression level in CA1 and Kcna1 has a higher expression level in CA3. Itpr1 encodes for intracellular channels that mediate calcium release from the endoplasmic reticulum (Yamada et. al, 1994), and Kcna1 encodes for voltage-gated delayed potassium channels (Curran et. al, 1991). The higher expression of Itpr1 and Kcna1 in CA3 is consistent with the higher tau value of CA3. In particular, on [GeneCards](https://www.genecards.org/cgi-bin/carddisp.pl?gene=KCNA1) for Kcna1, it stated that the N-terminus of the channel is associated with inactivation properties of the channel. So it is plausible that when having more Kcna1 expressed, there are more inhibitions of closure of the channel that drives higher tau. In terms of Itpr1 expression, it is possible that as a second messenger of the G-protein pathway, it facilitates the action potential production. Hence the time constant is smaller to allow for more action potential frequency. In conclusion, the observation of gene expression levels on both neurons could be contributing to the variation of time constant. 
To further extend the scope of this project, we also implemented a PCA model paired up with k-means clustering analysis to show whether the proposed electrophysiological features and differentially expressed genes could allow an unsupervised model to predict the region where this neuron is located. It turns out that both models, PCA and K-means clustering, showed an identical clustering of the samples. The number of samples in each clustering also matches with the number of labels for each region. Hence, it proved that the proposed electrophysiological features and genes are differ enough in two regions to allow the model to accurately distinguish them. The heatmap of the PCA factor value illustrated some extra features that distinguish the two neuron types. Aside from the two genes discussed above, Gabra1 and Scn8a also showed to have a good level of differences between them. Gabra1 is responsible for coding the subunit of GABA receptor (CA et. al, 2008), while Scn8a is responsible for giving instruction to build sodium channels (KM et. al, 2016). This implies that these two regions may have a great difference in inhibitory behaviors and one of them may have a relatively higher density of sodium channels. Yet, more biological experiments are needed to confirm this implication. 

Overall, the result of our analysis proved that while there isn’t a significant difference between the CA1 and CA3 regions, there is indeed a significant difference between the pyramidal cells in each region. Both the statistical tests and unsupervised clustering provided some insights into the major factors that contribute to the differences. 
There are some limitations of our analysis. The major one is we have a relatively small dataset which might make our findings affected by random noise. We have also looked into other databases but they either don’t provide a meta datasheet, or the format of the data was too complicated to handle with. For further analysis, we need more experiments looking at the differences in CA1 and CA3 electrophysiological properties to provide more recorded data. Another weakness would be the gene expression data. This analysis was performed under the assumption that all of these RNA sequences would be translated to protein. Yet this is not always the case in organisms. Hence it would be better next time if we can obtain data that directly measures the protein level. Lastly, the electrophysiological features and the gene expression were not measured from the same cells, so it is completely possible that our findings are not so relevant. In order to confirm our finding, following experiments could be a knockout experiment of the Kcna1 and Itpr1 gene in mice. In this way, we can confirm whether these two genes are the main contribution to the tau feature.
Additional potential experiments also include behavior related tasks performed by animal models, while the neuronal activity of CA1 and CA3 cells is recorded. In fact, Frank's lab has already released similar data. Yet, the format of the data was too complicated to handle so we didn’t include them in our analysis. This data will help us to visualize the differences between two neuron types when they are stimulated. For future reference, maybe an optogenetics experiment with direction manipulation of CA1 and CA3 neurons could help to explain the difference in functionalities as well.

```python
# Out of all the available CA1 observation, randomly pick 5 data points as trainning set and 2 as testing set. 
CA1_training_set = Combined_CA1_df.sample(n = 5, replace = False)
CA1_training_label = CA1_training_set['NeuronName']
CA1_training_feature = CA1_training_set.drop('NeuronName', axis = 1)

CA1_testing_set = Combined_CA1_df.drop(CA1_training_set.index)
CA1_testing_set = Combined_CA1_df.sample(n = 2, replace = False)
CA1_testing_label = CA1_testing_set['NeuronName']
CA1_testing_feature = CA1_testing_set.drop('NeuronName', axis = 1)
CA1_training_feature
```

```python
# Out of all the available CA3 observation, randomly pick 5 data points as trainning set and 2 as testing set. 
CA3_training_set = Combined_CA3_df.sample(n = 5, replace = False)
CA3_training_label = CA3_training_set['NeuronName']
CA3_training_feature = CA3_training_set.drop('NeuronName', axis = 1)

CA3_testing_set = Combined_CA3_df.drop(CA3_training_set.index)
CA3_testing_set = Combined_CA3_df.sample(n = 2, replace = False)
CA3_testing_label = CA3_testing_set['NeuronName']
CA3_testing_feature = CA3_testing_set.drop('NeuronName', axis = 1)
CA3_training_feature
```

```python
# Combine training set and testing set
Combined_training_feature = CA1_training_feature.append(CA3_training_feature)
Combined_training_label = CA1_training_label.append(CA3_training_label)

Combined_testing_feature = CA1_testing_feature.append(CA3_testing_feature)
Combined_testing_label = CA1_testing_label.append(CA3_testing_label)
Combined_testing_label
```

```python
# import the package for logistic regression and k-fold validation. Run the model and output the accuracy score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
clf = LogisticRegression()
clf.fit(Combined_training_feature, Combined_training_label)
clf.predict(Combined_testing_feature)

scores = cross_val_score(clf, Combined_training_feature, Combined_training_label, cv = 2)
print("Accuracy: %0.2f (+/- %0.2f)"% (scores.mean(), scores.std() *2))
```

The accuracy score turned out to be around 92%, meaning that the logisitic regression is performing well to predict the label (CA1/CA3 neurons). Yet it is noteworthy that the performance isn't too accurate at this point because the data set is too limited and this model is likely to be overfitted in reality.
