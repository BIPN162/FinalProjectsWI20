### I. c. Background & Previous Work
The hippocampal formation (HPF) is claimed to be involved in episodic, procedural, and spatial memory (Bird and Burgess 2008; Tulving and Marcowitsch 1998). A subarea of the HPF is the dentate gyrus (DG), which projects to CA3 area, which in turn projects to CA1 area (Strien et al. 2009). The DG has been associated with memory formation and is one of the only two human brain regions where neurogenesis occurs (Aimone et al. 2011, Barlow and Targum 2007). Many studies have attempted to link the generation of new neurons to memory plasticity, such as forgetting of old memories and discrimination of similar memories (Akers et al. 2014; Dery et al. 2013). With the DG having connections to other hippocampal formations (CA3/CA1), it will be interesting to study the association of neurogenesis and memory from a connectivity and gene expression level. 

The Allen Mouse Brain Connectivity Database is a map of neural connections in the mouse brain and provides us information regarding the projections of different brain area neuron groups. With this, we can discover or confirm whether one brain region projects to another. We can use this dataset to explore where DG neurons project to and to confirm that DG neurons indeed project to CA3. Examining whether the novel neurons in DG connect to CA3 can tell us if these newly generated neurons are physically involved in the HPF functional network. 

Additionally, comparing expression levels of putative neurogenesis-related genes in DG compared to other hippocampal areas can confirm theories about the uniqueness of neurogenesis, and can emphasize the role of synaptic development and neuronal rewiring of hippocampus-projecting DG neurons in memory functions. Some genes shown to be involved in neurogenesis are retinoic acid receptor ùõæ (Rxrg), fibroblast growth factor 3 (Fgfr3), and sex-determining region Y-box 2 (Sox2) (Aimone et al. 2011; Pozniak and Pleasure 2006; Barlow and Targum 2007). Genes involved in memory formation include glutamate ionotropic receptor NMDA Type Subunit 1 (grin1) and Discs Large MAGUK Scaffold Protein 3 (dlg3) (Lee 2011, Raymond 2004) The Allen Human Brain MicroArray database contains information regarding gene expression collected from 6 donor brains. We can search for a gene of interest and obtain its expression level in specific brain regions--specifically gene expression for the prior 5 mentioned genes in the DG, CA3, CA1 brain regions. 

Furthermore, it is commonly understood that patients with Alzheimer‚Äôs disease (AD) suffer from anterograde amnesia, or the ability to form new memories (Flicker et al. 1991). Dementia is another disorder characterized by impairment of memory. Since there is evidence that neurogenesis is related to several aspects of memory plasticity, it is worthwhile to explore whether neurogenesis processes and synaptic plasticity are affected in AD patients. RNAseq data is available for dementia patients in the Aging, Dementia and TBI datasets from the Allen Brain Institute. We can use this dataset in the same aspect as the MicroArray database and extract gene expression information for our 5 genes of interest from the hippocampus brain area. Ultimately from our data, we want to deduce how neurogenesis in the dentate gyrus may be related to memory plasticity.


### I. d. Hypothesis
If neurogenesis plays a role in memory plasticity and the dentate gyrus is a primary site of neurogenesis, then there should be an association between gene expression levels of neurogenesis and memory related genes with the highest expression levels of these genes being in the dentate gyrus when compared to other brain regions and in dementia patients, who are memory impaired.

The hippocampus is a brain region widely accepted to be involved in memory. The dentate gyrus is a substructure of the hippocampus and is a site of neurogenesis. Neurogenesis has been suggested to be involved with memory plasticity (Akers et al. 2014; Dery et al. 2013). Therefore, the dentate gyrus should have the highest gene expression levels of neurogenesis related genes when compared to other brain regions. Memory related gene expression would be elevated as well due to our assunmption that neurogenesis and memory are related. Gene expression would be expected to differ in the hippocampus of dmentia patients due to their impairments in memory, which may translate to impairments in genes responsible for memory.

## II. a. Dentate Gyrus' Connectivity to other Hippocampal Regions and Non-Hippocampal Regions in Mouse


Interested in exploring the connectivity of dentate gyrus neurons, we used RMA query functions to interact with the Mouse Brain Connectivity dataset (https://connectivity.brain-map.org/) from Allen Brain Institute to obtain the IDs of the experiments in which dentate gyrus neurons were labeled. 
The connectivity data was obtained by injecting Cre recombinase-dependent rAAV viral vectors into stereotaxically defined brain areas in mice expressing Cre recombinase in neurons that express specific gene promoters. Neurons that express these specific genes will be florescently labelled, and the florescence patterns are observed and recorded by two-photon imaging.

Through this database, we hope to find the 'popular regions' that DG granular neurons project to. In this way, we can learn about the brain regions which this area with active neurogenesis frequently talk to.

```python
# getting all experiment IDs that labelled neurons in the dentate gyrus.
dg_experiment_raw = requests.get('http://api.brain-map.org/api/v2/data/query.json?criteria=\
service::mouse_connectivity_injection_structure[injection_structures$eqDG][primary_structure_only$eqtrue]')

# the obtained file turned out to be a text file with all information included in a string. 
# I would therefore have to split the strings to extract the information I needed. 
dg_txt = dg_experiment_raw.text
mat = dg_txt.split(',')
counter = 0

for item in mat:
    if 'msg' in item:
        msg_position = counter
    counter += 1
mat_new = mat[msg_position:]
# mat_new is a relatively clean list containing the experiment IDs I am looking for, 
# but it still needs further cleaning.
```

The data I extracted included several transgenetically labelled neurons. Among all those transgenic lines, we want to look at which cell line labels cells mainly from the granular cell layer, as that is the layer the newly generated neurons reside.
According to http://gensat.org/cre.jsp, Grm2-Cre_MR90 and Rbp4-Cre_KL100 label granule cells in the dentate gyrus.

```python
# getting the experiment ids of the mouse transgenic lines
expt_id = []
counter2 = 0

for item in mat_new:
    if ("transgenic-line" in item):
        if 'Grm2-Cre_MR90' in item or 'Rbp4-Cre_KL100' in item:
            expt_id.append(counter2)
    counter2 += 1
    
real_id_ls = []
transgenic_line = []

for id_marker in expt_id:
    real_id_item = (mat_new[id_marker -1])
    real_id = real_id_item[-9:]
    real_id_ls.append(int(real_id))
    transgenic_txt = mat_new[id_marker]
    transgenic_line.append(transgenic_txt[18:])
    
experiment_df = pd.DataFrame({'EXPT_ID':real_id_ls, 'Transgenic_line': transgenic_line})
print ('These are the experiments I am interested in getting data from: ')
experiment_df
```

Interested in where the DG granular neurons project, we extracted the top five regions with highest projection density for each labelled neuron. We combined these projection densities and plotted the average projection densities for the areas appeared.

```python
# getting projection values and append to lists
big_conn_dataset = [conn_exp_1, conn_exp_2,conn_exp_3,conn_exp_4,conn_exp_5,conn_exp_6,\
                    conn_exp_7,conn_exp_8]
data_id = 0
top_ls = []
area_ls = []
density_ls = []

for data in big_conn_dataset:
    conn_exp_ls = data.text.split('\n')
    proj_value = []
    name_ls = []
    
    for item in conn_exp_ls:
        if 'projection-density' in item:
            proj_density_a = item.split('</')
            proj_density_b = proj_density_a[0].split('>')
            proj_value.append(float(proj_density_b[1]))
        if 'safe-name' in item:
            name_a = item.split('</')
            name_b = name_a[0].split('>')
            name_ls.append(name_b[1])
            
    summary = {'Name':name_ls, 'Projection density':proj_value}
    summary_df = pd.DataFrame(summary)
    summary_df_sorted = summary_df.sort_values(by= 'Projection density',ascending = False)
    summary_df_reset = summary_df_sorted.set_index('Name')
    top_ls.append(summary_df_reset.head())
    data_id += 1
    
counter = 0
y = {}

for session in top_ls:
    for item2 in session.iterrows():
        if item2[0] not in area_ls:
            area_ls.append(item2[0])
            y[counter] = []
            y[counter].append(item2[1].values)
            counter += 1
        else:
            indx = area_ls.index(item2[0])
            y[indx].append(item2[1].values)
mean_expression = []
std_expression = []

for i in list(range(0,len(y))):
    mean = float(sum(y[i])/len(y[i]))
    std = np.std(y[i])
    mean_expression.append(mean)
    std_expression.append(std)
```

Next, since we got all the data for the highest projection values from DG to other brain regions, we want to plot the projection values in a bargraph and compare.

```python
# plot the projection density values of each brain region into a bargraph
plt.bar(area_ls, mean_expression,yerr = std_expression)
plt.xticks(area_ls, rotation = 90)
plt.ylabel('Projection Density')
plt.title('Figure 1. Average projection density of Regions Most Densely Targeted by DG Granular Neurons in Eight Experiment Sessions')
plt.show()
```

Figure 1 showed that CA3, CA2, and dentate gyrus itself are among the most connected areas by DG granular neurons. Error bars were not shown for the first four areas, because only one neuron examined projected to those areas, which also meant that DG projections to these four areas were not as frequent as the other six areas. This figure clearly showed that DG granular layer, the layer with neurogenic activities, has strong projections to other hippocampal areas in the HPF memory network.

In the data collection process of this microarray database, researchers first obtained tissue with location specified by DTI/MRI imaging, and then purified the tissue for microarray analysis. Purified samples were sent to a third-party company where probes with thousands of differnt mRNA sequences were used to bind and measure the concentration of mRNA in each tissue sample, and the data was quantified and normalized.

Although we do not see significant difference in PCA, we still want to make more statistical analysis about gene expressions in the four brain regions. To compare gene expressions across different areas, we want to use one-way ANOVA, which tests for the significance of an independent variable in a sample, to test the relationships between genes studied. The p-values are shown in the output below. For these ANOVA tests, the null hypothesis is: there is no statistical difference in gene expression levels across all brain areas. 

```python
# doing one way ANOVA test for gene expression significances in different brain regions
fgfr3_anova, fgfr3_anova_p = sp.stats.f_oneway(exp1_data[0], exp1_data[1], exp1_data[2], exp1_data[3])
rxrg_anova, rxrg_anova_p = sp.stats.f_oneway(exp2_data[0], exp2_data[1], exp2_data[2], exp2_data[3])
sox2_anova, sox2_anova_p = sp.stats.f_oneway(exp3_data[0], exp3_data[1], exp3_data[2], exp3_data[3])
grin1_anova, grin1_anova_p = sp.stats.f_oneway(exp4_data[0], exp4_data[1], exp4_data[2], exp4_data[3])
dlg3_anova, dlg3_anova_p = sp.stats.f_oneway(exp5_data[0], exp5_data[1], exp5_data[2], exp5_data[3])

print (('ANOVA for FGFR3 in DG, CA3, CA1, PoG is %f, p-value is %f') % (fgfr3_anova, fgfr3_anova_p))
print (('ANOVA for RXRG in DG, CA3, CA1, PoG is %f, p-value is %f') % (rxrg_anova, rxrg_anova_p))
print (('ANOVA for SOX2 in DG, CA3, CA1, PoG is %f, p-value is %f') % (sox2_anova, sox2_anova_p))
print (('ANOVA for GRIN1 in DG, CA3, CA1, PoG is %f, p-value is %f') % (grin1_anova, grin1_anova_p))
print (('ANOVA for DLG3 in DG, CA3, CA1, PoG is %f, p-value is %f') % (dlg3_anova, dlg3_anova_p))
```

The p value for FGFR3 gene expression is bigger than 0.05, which means that we do not have enough evidence to conclude that FGFR3 expression levels are different across all four brain areas. 

The p-value for the other four genes are smaller than 0.05, meaning that gene expressions in some areas are different from others. However, in order to determine which brain areas show different gene expressions from others, we have to conduct a post-hoc test. We chose to use Tukey Honest Significance Test (or Tukey HSD). 

```python
from statsmodels.stats.multicomp import (pairwise_tukeyhsd,MultiComparison)

# Organizing data into groups in a dataframe
for exp_dt in data_loop:
    df = pd.DataFrame()
    df['treatment1'] = exp_dt[0]
    df['treatment2'] = exp_dt[1]
    df['treatment3'] = exp_dt[2]
    df['treatment4'] = exp_dt[3]
    stacked_data = df.stack().reset_index()
    stacked_data = stacked_data.rename(columns={'level_0': 'id','level_1': 'treatment',0:'result'})

# Setting up data for comparison
    MultiComp = MultiComparison(stacked_data['result'],stacked_data['treatment'])

# Print the test result
    print(MultiComp.tukeyhsd().summary(), '\n')
```

The tables above showed results of the Tukey test. From top to bottom, each table showed comparison of mean difference of FGFR3, RXRG, SOX2, GRIN1, and DLG3 expression, respectively.

From the results above, there are two interesting observations. The two plasticity-related genes (GRIN1, DLG3) show higher expression in some memory-related brain areas than the non-memory related control (post-central gyrus) (see sub-table 4 and 5, comparison of treatment 1-3 against treatment 4). Both GRIN1 and DLG3 expressions are higher in DG than in postcentral gyrus.

Additionally, according to subtable 3, all HPF areas had lower SOX2 expression than the post-central gyrus.

```python
# to get gene expression levels from Age, Dementia, TBI study service
sox2_exp = []
sox2_z = []
rxrg_exp = []
rxrg_z = []
fgfr3_exp = []
fgfr3_z = []
grin1_exp = []
grin1_z = []
dlg3_exp = []
dlg3_z = []

# got probe ids from the website:
sox2_id = 499315403
rxrg_id = 499307888
fgfr3_id = 499315803
grin1_id = 499329736
dlg3_id = 499353433

for donor in donor_id:
    
    # sox2 expressions
    sox2_results = requests.get('%sservice::tbi_expression[donors$eq\'%i\'][structures$eq\'%s\']\
    [probes$eq\'%i\']' % (service, donor, structure_id, sox2_id)).json()
    
    if sox2_results['msg']['probes'][0]['expression_level'] == []:
        sox2_exp.append(None) 
        sox2_z.append(None)
    else:
        sox2_exp.append(float(sox2_results['msg']['probes'][0]['expression_level'][0]))
        sox2_z.append(float(sox2_results['msg']['probes'][0]['z-score'][0]))
        
    # rxrg expressions
    rxrg_results = requests.get('%sservice::tbi_expression[donors$eq\'%i\'][structures$eq\'%s\']\
    [probes$eq\'%i\']' % (service, donor, structure_id, rxrg_id)).json()
    
    if rxrg_results['msg']['probes'][0]['expression_level'] == []:
        rxrg_exp.append(None) 
        rxrg_z.append(None)
    else:
        rxrg_exp.append(float(rxrg_results['msg']['probes'][0]['expression_level'][0]))
        rxrg_z.append(float(rxrg_results['msg']['probes'][0]['z-score'][0]))
        
    # fgfr3 expressions
    fgfr3_results = requests.get('%sservice::tbi_expression[donors$eq\'%i\'][structures$eq\'%s\']\
    [probes$eq\'%i\']' % (service, donor, structure_id, fgfr3_id)).json()
    
    if fgfr3_results['msg']['probes'][0]['expression_level'] == []:
        fgfr3_exp.append(None) 
        fgfr3_z.append(None)
    else:
        fgfr3_exp.append(float(fgfr3_results['msg']['probes'][0]['expression_level'][0]))
        fgfr3_z.append(float(fgfr3_results['msg']['probes'][0]['z-score'][0]))
        
    # grin1 expressions
    grin1_results = requests.get('%sservice::tbi_expression[donors$eq\'%i\'][structures$eq\'%s\']\
    [probes$eq\'%i\']' % (service, donor, structure_id, grin1_id)).json()
    
    if grin1_results['msg']['probes'][0]['expression_level'] == []:
        grin1_exp.append(None) 
        grin1_z.append(None)
    else:
        grin1_exp.append(float(grin1_results['msg']['probes'][0]['expression_level'][0]))
        grin1_z.append(float(grin1_results['msg']['probes'][0]['z-score'][0]))
        
    # dlg3 expressions
    dlg3_results = requests.get('%sservice::tbi_expression[donors$eq\'%i\'][structures$eq\'%s\']\
    [probes$eq\'%i\']' % (service, donor, structure_id, dlg3_id)).json()
    
    if dlg3_results['msg']['probes'][0]['expression_level'] == []:
        dlg3_exp.append(None) 
        dlg3_z.append(None)
    else:
        dlg3_exp.append(float(dlg3_results['msg']['probes'][0]['expression_level'][0]))
        dlg3_z.append(float(dlg3_results['msg']['probes'][0]['z-score'][0]))
```

Then we will combine all the retrieved data into a single dataframe, for the convenience of plotting.

```python
# combine all the data into one dataframe (without n/a values)
tbi_df = pd.DataFrame({'donor_id': donor_id, 'donor_age': donor_age, \
                            'disease_state': disease_state, 'sox2_exp': sox2_exp, \
                           'sox2_zscore': sox2_z, 'rxrg_exp': rxrg_exp, 'rxrg_zscore': rxrg_z, \
                           'fgfr3_exp': fgfr3_exp, 'fgfr3_zscore': fgfr3_z, 'grin1_exp': \
                       grin1_exp, 'grin1_zscore': grin1_z, 'dlg3_exp': dlg3_exp, \
                       'dlg3_zscore': dlg3_z}).dropna()
tbi_df.head()
```

After getting the dataframe containing all the data we want, next we will plot the gene expression values and the corresponding z scores to visualize and compare the gene expressions between subjects with and without dementia.

```python
# plot boxplots of all the gene expressions with seaborn
sns.set_style('whitegrid')

fig, ax = plt.subplots(figsize = (10, 8))

# plot gene expression values
# melt the datframe into a long form so it's easier to plot with seaborn
tbi_exp_long = pd.melt(tbi_df, id_vars = ['disease_state'], value_vars = 
                       ['sox2_exp', 'rxrg_exp', 'fgfr3_exp', 'grin1_exp', 'dlg3_exp'], var_name = 'genes', value_name \
                       = 'expression values/fpkm')
exp_plot = sns.boxplot(x = 'genes', y = 'expression values/fpkm', ax = ax, data = tbi_exp_long, \
                 hue = 'disease_state', palette = 'colorblind')
ax.set_title('Figure 4. Gene Expression Values in Dementia vs. Control Subjects')

fig.show()
```

From the boxplots above, it seems that there are no significant differences of the gene expressions in dementia and none-dementia subjects. Although we have a visualization of the data, we still want to confirm what we see is statistically true. In the code below, we will perform Mann-Whitney U test to see the significance of the gene expression comparisons. Mann-Whitney U test is chosen because based on the flow chart of choosing the right statistic test provided in the course notebook 'Statistics', the two samples (dementia, no dementia) are selected from the same population but not paired.

```python
# to select and reassign data based on disease state and genes
dementia_df = tbi_df[tbi_df['disease_state'] == 'Dementia']
control_df = tbi_df[tbi_df['disease_state'] == 'No Dementia']
print(('The length of dementia_df is %i, and the length of control_df is %i.') % (len(dementia_df), len(control_df)))

sox2_dementia = dementia_df[['sox2_exp']]
rxrg_dementia = dementia_df[['rxrg_exp']]
fgfr3_dementia = dementia_df[['fgfr3_exp']]
grin1_dementia = dementia_df[['grin1_exp']]
dlg3_dementia = dementia_df[['dlg3_exp']]

sox2_control = control_df[['sox2_exp']]
rxrg_control = control_df[['rxrg_exp']]
fgfr3_control = control_df[['fgfr3_exp']]
grin1_control = control_df[['grin1_exp']]
dlg3_control = control_df[['dlg3_exp']]

# doing Mann-Whitney U tests for all gene expressions
sox2_zstat,sox2_p = stats.mannwhitneyu(sox2_dementia, sox2_control)
rxrg_zstat, rxrg_p = stats.mannwhitneyu(rxrg_dementia, rxrg_control)
fgfr3_zstat, fgfr3_p = stats.mannwhitneyu(fgfr3_dementia, fgfr3_control)
grin1_zstat, grin1_p = stats.mannwhitneyu(grin1_dementia, grin1_control)
dlg3_zstat, dlg3_p = stats.mannwhitneyu(dlg3_dementia, dlg3_control)

print('\nMann-Whitney U test results for dementia vs. control:')
print(('SOX2 expression z score is %f, p-value is %f.') % (sox2_zstat, sox2_p))
print(('RXRG expression z score is %f, p-value is %f.') % (rxrg_zstat, rxrg_p))
print(('FGFR3 expression z score is %f, p-value is %f.') % (fgfr3_zstat, fgfr3_p))
print(('GRIN1 expression z score is %f, p-value is %f.') % (grin1_zstat, grin1_p))
print(('DLG3 expression z score is %f, p-value is %f.') % (dlg3_zstat, dlg3_p))

if sox2_p < 0.05:
    print('\nSOX2 expression is significantly different!')
if rxrg_p < 0.05:
    print('\nRXRG expression is significantly different!')
if fgfr3_p < 0.05:
    print('\nFGFR3 expression is significantly different!')
if grin1_p < 0.05:
    print('\nGRIN1 expression is significantly different!')
if dlg3_p < 0.05:
    print('\nDLG3 expression is significantly different!')
```

From the Mann-Whitney U test above, there are no significant differences in gene expressions except for RXRG gene. It is difficult to draw conclusion from the t test above to say whether the regional gene expressions are indeed different or not. One reason for the diffculty is that there are many variables (or dimensions) that cannot be combined together with a simple t test. However, PCA can help reduce dimensions of a dataset and give a statistical analysis that combines all the different variables. Therefore by performing PCA, we can generalize if all the gene expressions together are significantly different between dementia and control groups or not.

```python
# to reassign normalized data and drop 'disease_state' column
tbi_zscore = tbi_df[['disease_state', 'sox2_zscore', 'rxrg_zscore', 'fgfr3_zscore', 'grin1_zscore', 'dlg3_zscore']]
tbi_zscore_norm = tbi_zscore.drop('disease_state', axis = 1)

# to perform and plot PCA
pca = PCA(n_components = 2)
tbi_zscore_norm_pca = pca.fit_transform(tbi_zscore_norm)

tbi_zscore['PC1'] = tbi_zscore_norm_pca[:, 0]
tbi_zscore['PC2'] = tbi_zscore_norm_pca[:, 1]
tbi_zscore

sns.lmplot('PC1', 'PC2', data = tbi_zscore, hue = 'disease_state', fit_reg = False)
plt.title('Figure 5. PCA of Hippocampal Gene Expressions in Dementia vs. Control')
plt.show()

# see what ratios of variance are explained by the principle components
print((' Explained variance ratio for PC1 is %f and for PC2 is %f.') % (pca.explained_variance_ratio_[0], pca.explained_variance_ratio_[1]))
```

From the above PCA, it seems that there is no significant seperation between the two clusters representing disease state, suggesting that the hippocampal gene expression values are not significantly different in subjects with dementia and without dementia. Therefore, all the above analysis are not enough to confirm our hypothesis that gene expressions that mark neurogenesis and memory function might be differentially expressed in the hippocampus region between dementia patients and controls.

## III. Conclusion & Discussion

Since we are interested in understaning the possible role of neurogenesis in memory function in hippocampus, our data science approach offers a method to look into the correlation of hippocampal neurogenesis to neuronal plasticity, with various data acquired from Allen Brain Atlas, such as Mouse Connectivity Atlas, Human Microarray dataset, and Aging, Dementia, TBI dataset. However, we are not able to determine a correlation between hippocampal neurogenesis expressions and hippocampal neuronal plasticity expressions. 

First, we looked at the Mouse Connectivity dataset to see the connections of dentate gyrus to other brain regions, hoping to confirm the hippocampal memory network (dentate gyrus-CA3-CA1) proposed by previous research (Aimone et al., 2011). Indeed, from Fig. 1 we see high projection densities from dendate gyrus to dendate gyrus itself, and also to CA1 and CA3 regions of the hippocampus. This data suggests that the memory network is present in our dataset. Since the granular layer of dentate gyrus is proposed to be where neurogenesis happens, the data indicates that maybe neurogenesis is associated with memory function in CA1 and CA3. 

Second, to see if neurogenesis is related to memory function (specifically neuronal plasticity) from gene expression analysis, we looked into the Human Microarray dataset. We selected FGFR3, SOX2 and RXRG for neurogenesis expression, and GRIN1 and DLG3 for plasticity expression. We also selected four brain regions to look at, DG, CA1, CA3 and PoG. PoG is part of the somatosensory cortex, and it is not likely to highly expression memory function; thus PoG is used as a control for plasticity expression. From the boxplots of Fig.2, we see that RXRG expression is higher in CA3 and GRIN1 & DlG3 are lower in PoG, but there is no significant difference in FGFR3 expression across the four brain regions. Additionally, these results are further confirmed by ANOVA test, which only sees no significance in FGFR3 expression. Since we do see both plasticity expressions lower in PoG, this confirms our hypothesis that PoG does not likely have memory-related functions. Interestingly, from both Fig.2 and Tukey test, SOX2 expression is higher in PoG region than the hippocampal regions. This result was unexpected, because we learned that Sox2 promotes differentiation of progenitor cells from many existing literature (Aimone et al. 2011), and postcentral gyrus is not one of the areas with active neurogenesis. However, as we searched for further discussions on this topic, we found several other studies proposing an inhibitory role of Sox2 on neurogenesis (Bylund et al. 2003; Oosterveen et al. 2013), and it was actually other proteins in the Sox family, such as Sox21 that promote progenitor differentiation into neurons. Such conflicting findings indicated that there may exist more complicated gene regulation mechanisms. Adding the confounding results, PCA of Fig.3 demonstrates no significant seperations of different brain region clusters. This result indicates that the neurogenesis and plasticity expressions might not actually be very different in hippocampal and non-hippocampal regions. In short, we cannot validate that hippocampal neurogenesis correlates to plasticity from the microarray dataset.

Last, since dementia is a neurodegenerative disease and manifests in impaired neurogenesis and memory function (A. Burns, 2009), we want to see if we can find a correlation betweeen neurogenesis and plasticity from the Age, Dementia, TBI Study dataset. In Fig.4, we compared neurogenesis and plasticity expressions in hippocampus of dementia and non-dementia human subjects. Fig.4 demonstrates that there is no significant difference between dementia patients and healthy controls, suggesting that these genes might not be correlated with memory function. Moreover, the result also could suggest that we might not be looking at the right genes that show a difference in memory performance between dementia patients and controls, thereby not giving us any clues about the connection between neurogenesis and memory function in hippocampus. 

One limitation in our project is that only mouse data is avaiable for the connectivity database. Although the mouse and human brains have similar structures, species difference still cannot transfer the results from mouse to human. Furthermore, sample size is a clear limitation in our data collection. The Allen Human Microarray database only contains information from six donor brains. A small sample size consequently reduces the power of statistics and analysis, giving us information not representative of the population. This is especially true for PCA, where larger sample may have a more representative clustering. Moreover, dementia is a very broad category of disease, which includes Alzheimer's. The cuasation of dementia in our dataset is not clear, thus we do not know what genes or to what extent, the genes we studied affect neurogenesis or memory. 

Further studies may include electrophysiology and morphology to better understand neuronal plasticity. Representing levels of synaptic plasticity using gene expression is a great way to quantify such a concept. However, more comprehensive measurement can be obtained by electrophysiology or morphology. We would like to extend our project by measuring electrophysiological properties of pyramidal neurons in DG, CA1, CA3, and PoG. We would need recordings of the neurons of interest from healthy human subjects during resting periods, in order to compare the neurons‚Äô intrinsic properties. Another direction for further study would be measuring the rate of increase/decrease of the number of neurons in brain areas. 
Studying neurogenesis levels through gene expressions is, ultimately, an indirect way of measurement. The regulation of genes responsible for mitosis and differentiation can be very complex to account for. To understand if changes in neurogenesis is correlated to changes in memory functions, we could look at the rate of neuronal regeneration in the dentate gyrus in healthy subjects and AD patients. We could mark actively-dividing cells in the DG and calculate the intensity of fluorescence, and then compare the intensity across AD patients and control.
